```python
import networkx as nx
import matplotlib.pyplot as plt

def create_cayley_tree(k, n):
    """Creates a Cayley tree with degree k and n nodes."""
    return nx.full_rary_tree(k, n)

# Create a Cayley tree with degree 3 and 13 nodes
tree = create_cayley_tree(3, 4)

# Draw the tree
pos = nx.spring_layout(tree)  # Adjust layout as needed
nx.draw(tree, pos, with_labels=True)
plt.show()
```

This is the smallest possible Cayley tree, where the number of steps $n = 1$.
