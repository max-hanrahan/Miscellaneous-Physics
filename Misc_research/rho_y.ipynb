```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
names = ['ATOMS', 'id', 'type', 'x', 'y', 'z', 'vx', 'vy', 'vz']
```

```python
Natoms = 3286
```

```python
def density_profile(config):
    row_number = (Natoms + 9)* config + 9
    df = pd.read_csv('coords-vel.xyz', skiprows = row_number, nrows = Natoms, 
                     delim_whitespace = True, header = None, names = names)
    df_hist = plt.hist(df.vx, bins = 200, histtype = 'step', color=plt.cm.viridis(config/20), alpha = 0.5)
    return df_hist
```

```python
# indicate wave dispersing over time using the colormap
for i in range(20):
    density_profile(i)
plt.ylabel(r"Un-normalized $v(x)$")
plt.xlabel(r"$x$")
```
