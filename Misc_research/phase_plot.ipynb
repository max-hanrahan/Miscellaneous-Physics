# Phase diagrams for different $\epsilon_{AB}$ at $P=1$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
T_list = np.arange(1.00, 4.00, 0.10)
path_to_file = "home/SGNPT_epsAB/95/"
```

```python
# read in the data without having to manually write their names
def read_dat(filename):
    df = pd.read_csv(filename, header = 0,  delim_whitespace=True)
    df_modified = df[df.columns[:-1]] # get rid of the leading #
    df_modified.columns = df.columns[1:] # reset the columns
    return df_modified
```

```python
# function to filter out the first half of the dataset
def select_second_half(df):
    half = df.Step[(len(df)) // 2]
    df = df[df.Step>half]
    return df
```

```python
path_to_file = "home/SGNPT_epsAB/95/"
phi_1 = []
for T in T_list:
    file = f"{path_to_file}/T{T:.2f}/thermo.dat"
    df = select_second_half(read_dat(file))
    df['Density'] = pd.to_numeric(df['Density'])
    phi_1.append(np.mean(df['Density']))    
    # plt.plot(df.Density)
phi_1 = np.array(phi_1)
plt.style.use('bmh')
plt.scatter(phi_1, T_list)
# plt.scatter(1-phi_1, T_list)
```

```python
path_to_file = "home/SGNPT_epsAB/90/"
phi_2 = []
for T in T_list:
    file = f"{path_to_file}/T{T:.2f}/thermo.dat"
    df = select_second_half(read_dat(file))
    df['Density'] = pd.to_numeric(df['Density'])
    phi_2.append(np.mean(df['Density']))    
phi_2 = np.array(phi_2)
plt.style.use('bmh')
plt.scatter(phi_2, T_list)
# plt.scatter(1-phi_1, T_list)
```

```python
plt.plot(phi_1, T_list, label = 'eps = 0.95')
plt.plot(phi_2, T_list, label = 'eps = 0.90')
plt.xlabel("density")
plt.ylabel("Temp")
plt.legend()
```
