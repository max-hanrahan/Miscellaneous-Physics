## 1: 

Verify that Hamilton's Principal function for the  1d H.O. with kinetic energy $K = \frac{1}{2}m \dot x^2$ and potential energy $V(x) = m \omega^2 x^2/2$ is:

As we showed last week, the Lagrangian for such a system is given by:

$$
\mathcal{L}=\frac{1}{2} m \omega^2\left[A^2\left(\sin ^2 \omega t+\cos ^2 \omega t\right)+B^2\left(\cos ^2 \omega t-\sin ^2 \omega t\right)-4 A B \sin \omega t \omega \cos k t\right],
$$

where

$$
A = \frac{x_i \sin(\omega t_f) - x_f \sin(\omega t_i)}{\sin[\omega(t_f - t_i)]},
$$

$$
B = \frac{-x_i \cos(\omega t_f) + x_f \cos(\omega t_i)}{\sin[\omega(t_f - t_i)]}.
$$


So, we can use

$$
\begin{aligned}
R(x_i t_i, x_f t_f) &= \int\mathcal{L}dt \\
&= \int_{t_i}^{t_f}\frac{1}{2} m \omega^2\left[A^2\left(\sin ^2 \omega t+\cos ^2 \omega t\right)+B^2\left(\cos ^2 \omega t-\sin ^2 \omega t\right)-4 A B \sin \omega t \omega \cos k t\right]dt,
\end{aligned}
$$

and we can quote this result right away, because this is the same integral we computed on problem 5 of HW8. 

Its result is indeed 

$$
\boxed{\frac{m \omega}{2 \sin \omega T} \left[ (x_f^2 + x_i^2) \cos \omega T - 2 x_f x_i \right], \quad T = t_f - t_i}
$$

which matches the expression we sought to show. 

## 2:

Illustrating Gutzweiller's second modification of the Van Vleck Propagator, compute it for free motion of a particle of mass $m$ on the positive $x$-axis with a reflecting wall at $x=0$. Compare with the exact result of 12.52. 

For reference:

$$
\mathcal G_\text{HS}^{(+)} (x_f \, t_f, x_i \, t_i) = \sqrt{\frac{m}{2 \pi i \hbar (t_f - t_i)}} \left\{\exp \left[i \frac{m(x_f - x_i)^2}{2 \hbar (t_f -t_i)}\right] - \exp \left[i \frac{m(x_f + x_i)^2}{2 \hbar (t_f -t_i)}\right]\right\} \tag{12.52}
$$


Equation 13.31 is a general formula for the VVG formula. However, there are a few distinctions that we should bear in mind that prevent us from simply plugging and chugging. 

If a _classical_ particle in this system moves from any point A to point B, it could have gotten there in one of two ways: either directly, or by bouncing off the wall. If it got there directly, the propagator associated with that path is the same one for free space:

$$
\mathcal{G}_\text{free}(x_f t_f, x_i t_i) = \sqrt{\frac{m}{(t_f - t_i)2 \pi i \hbar}} \left(\exp\left\{\frac{im(x_f - x_i)^2}{2\hbar (t_f - t_i)} \right\}\right). \\
$$

Conceptually speaking, Gutzweiller's second modification is to sum over all possible trajectories, so this is the first one in our sum. The second (and only other trajectory) would be bouncing off the hard wall.

This is why we need to slightly modify 13.31 in the book: the $\pi/2$ there is only for _soft_ potentials. Hard potentials reflect according to a phase factor $\pi$. As such, our alternate path has the following propagator:

$$
\mathcal{G_\text{bounce}} = \left(\frac{1}{2 \pi i \hbar}\right)^{d/2} \sqrt{|\mathcal{D}_R^{(j)}|} \exp\left\{\frac{iR^{(j)}(r_ft_f, r_it_i)}{\hbar} - i\pi \nu_D^{(j)}\right\},
$$

which we can simplify greatly. We know the number of dimensions $d = 1$, and the Van Vleck determinant $\mathcal{D}_R$ do not change from what we found in the textbook for the case of free space. What does change is Hamilton's principal function $R^{(j)}$, as well as the number of congugate points $\nu_D$, which is one for this wall -- cause there's only one way to experience a singularity on your journey from point A to point B, and it is by bouncing off the wall. 

The new $R^{(j)}$ is computed from the Lagrangian. $\mathcal{L}$ is the same here as it was for free space,   

$$
\mathcal L(x, \dot{x}) = \frac{1}{2} m \dot{x}^2 
$$

but since we're considering only the reflected path, the particle covered a different distance to get from point A to point B in the same stretch of time. Previously we had

$$
\dot x = \frac{x_f - x_i}{t_f - t_i}, 
$$

but since we reflect, we have

$$
\dot x = \frac{x_f - (-x_i)}{t_f - t_i} = \frac{x_f + x_i}{t_f - t_i}, 
$$

and our Lagrangian becomes 

$$
\mathcal L(x, \dot{x}) = \frac{1}{2} m \dot{x}^2  = \frac{1}{2} m\left(\frac{x_f + x_i}{t_f - t_i}\right)^2, 
$$

which we can readily use to find our H. principal value:

$$
R  = \int_{t_{i}}^{t_f}\mathcal L(x, \dot{x}) dt = \frac{1}{2} m \frac{(x_f + x_i)^2}{t_f - t_i}
$$

Therefore, our second term in the VVGP reduces to


$$
\begin{aligned}
\mathcal{G_\text{bounce}} &= \left(\frac{1}{2 \pi i \hbar}\right)^{1/2} \sqrt{\frac{m}{t_f - t_i}} \exp\left\{\frac{im(x_f + x_i)^2}{2\hbar (t_f - t_i)} - i\pi \right\}  \\
=&\ \sqrt{\frac{m}{(t_f - t_i)2 \pi i \hbar}} \exp\left\{\frac{im(x_f + x_i)^2}{2\hbar (t_f - t_i)} - i\pi \right\} \\
=&\ \sqrt{\frac{m}{(t_f - t_i)2 \pi i \hbar}} \left(-\exp\left\{\frac{im(x_f + x_i)^2}{2\hbar (t_f - t_i)}  \right\}\right). \\
\end{aligned}
$$

Now that we have both terms, we simply collect them:


$$
\begin{aligned}
\mathcal{G}_{VVG} &= \mathcal{G}_\text{Free} + \mathcal{G_\text{bounce}}\\
 &= \sqrt{\frac{m}{(t_f - t_i)2 \pi i \hbar}} \left(\exp\left\{\frac{im(x_f - x_i)^2}{2\hbar (t_f - t_i)} \right\}\right) + \sqrt{\frac{m}{(t_f - t_i)2 \pi i \hbar}} \left(-\exp\left\{\frac{im(x_f + x_i)^2}{2\hbar (t_f - t_i)}  \right\}\right) \\
 &= \sqrt{\frac{m}{2 \pi i \hbar (t_f - t_i)}} \left\{\exp \left[i \frac{m(x_f - x_i)^2}{2 \hbar (t_f -t_i)}\right] - \exp \left[i \frac{m(x_f + x_i)^2}{2 \hbar (t_f -t_i)}\right]\right\}
\end{aligned}
$$ 

## 3:

A particle of mass $m$ moves on the $x$ axis in the potential

$$
V(x) = -\beta x.
$$

Compute the Van Vleck-Gutzwiller Propagator $\mathcal{G}_\text{VVG}(x_f, x_i; E)$ for $0 < x_i< x_f$ and $E > - \beta x_1$. 

Let's start from 13.94 in the book, slightly rephrased to accomodate multiple paths:

$$
\mathcal{G}_{VVG}^{d = 1}(x_f, x_i, E) = \sum_{\text{paths}} \frac{1}{i \hbar \sqrt{|\dot x_f \dot x_i|}} \exp\big\{i \left[\frac{1}{\hbar}S(x_f, x_i; E) - \mu \frac{\pi}{2}\right]\big\}.
$$

Regardless of path, the following is true:

$$
E = \frac{m\dot x_i^2}{2} - \beta x_i \implies \dot x_i = \sqrt{\frac{2}{m} (E+ \beta x_i)},
$$

and similarly

$$
\dot x_f = \sqrt{\frac{2}{m} (E+ \beta x_f)}, 
$$

so we can evaluate the pre-exponential factor of the sum right off the bat:

$$
\begin{aligned}
\frac{1}{i \hbar \sqrt{|\dot x_f \dot x_i|}}&= \frac{1}{i \hbar \sqrt{|\sqrt{\frac{2}{m} (E+ \beta x_f)} \sqrt{\frac{2}{m} (E+ \beta x_i)}|}} \\
&= \frac{1}{i \hbar} \sqrt{\frac{m}{2}} (E^2 + \beta E (x_i + x_f) + \beta^2 x_i x_f)^{-1/4}
\end{aligned}
$$

What _does_ depend on path is the action. One path is simply straight from $x_i$ to $x_f$, and the other goes through the turning points. 

The straightforward path has action

$$
\begin{aligned}
S_1(E) &= \int_{x_i}^{x_f} p dx = \int_{x_i}^{x_f} \sqrt{2m(E + \beta x)}dx \\
&= \frac{2}{3\beta} \sqrt{2m }(E+ \beta x)^{3/2} \big|_{x_i}^{x_f} \\
&= \frac{2}{3\beta} \sqrt{2m }(E+ \beta x_f)^{3/2} - \frac{2}{3\beta} \sqrt{2m }(E+ \beta x_i)^{3/2} \\
&= \frac{2 \sqrt{2m}}{3 \beta} \left[(E + \beta x_f)^{3/2} - (E+ \beta x_i)^{3/2} \right].
\end{aligned}
$$

Now for the second path, we must go through the turning point $x_1$:

$$
-\beta x_1 = E \implies x_1 = -\frac{E}{\beta}, 
$$

so then the action from $x_i$ to $x_1$ is simply

$$
\begin{aligned}
S_{i \to 1}(E) &= \int_{x_1}^{x_i} p dx = \int_{x_1}^{x_i} \sqrt{2m(E + \beta x)}dx \\
&= \frac{2}{3\beta} \sqrt{2m }(E+ \beta x)^{3/2} \big|_{x_1}^{x_i} \\
&= \frac{2}{3\beta} \sqrt{2m }(E+ \beta x_i)^{3/2} - \frac{2}{3\beta} \sqrt{2m }(E+ \beta x_1)^{3/2}. \\
&= \frac{2}{3\beta} \sqrt{2m }(E+ \beta x_i)^{3/2} - \frac{2}{3\beta} \sqrt{2m }(E-E)^{3/2}. \\
&= \frac{2}{3\beta} \sqrt{2m }(E+ \beta x_i)^{3/2}.
\end{aligned}
$$

Now similarly the action of the path from $x_i$ to $x_1$ to $x_i$ is twice the action of $x_i$ to $x_1$:

$$
\begin{aligned}
S_2(x_f, x_i; E) &= 2S_{i \to 1}(x_f, x_i; E) + S_1(x_f, x_i; E) \\
&= \frac{2 \sqrt{2m}}{3 \beta} \left[(E + \beta x_f)^{3/2} - (E+ \beta x_i)^{3/2} + 2(E+ \beta x_i)^{3/2}\right] \\
&= \frac{2 \sqrt{2m}}{3 \beta} \left[(E + \beta x_f)^{3/2} + (E+ \beta x_i)^{3/2} \right].
\end{aligned}
$$




We are now ready to sum the paths. The first path has the Maslov index $\mu_1 = 0$ and then path 2 picks up a phase of $\frac{-i \pi }{2}$ (i.e., $\mu_2 = 1$) upon hitting the soft wall. 

Our propagator is therefore:



$$
\begin{aligned}
\mathcal{G}_{VVG}(x_f, x_i, E)= \frac{1}{i \hbar} \sqrt{\frac{m}{2}} 
&\left(E^2 + \beta E (x_i + x_f) + \beta^2 x_i x_f\right)^{-1/4} \\
&\times \Bigg[ \exp\left\{ 
    i \left( \frac{2\sqrt{2m}}{3\beta\hbar} (E + \beta x_f)^{3/2} - (E + \beta x_i)^{3/2} \right)
\right\} \\
&\quad + \exp\left\{
    i \left( \frac{2\sqrt{2m}}{3\beta\hbar} 
    \left[(E + \beta x_f)^{3/2} + (E + \beta x_i)^{3/2} \right] 
    - \frac{\pi}{2} \right)
\right\} \Bigg]
\end{aligned}
$$


## 4:

Show that in the special case of one dimensional free mtion, the energy-dependent Van Vleck-Gutzweiler propagator in 13.94 is identical with the exact propagator in 12.78. 


For reference, 13.94:

$$
\mathcal{G}_{VVG}^{d = 1}(x_f, x_i, E) = \frac{1}{i \hbar \sqrt{|\dot x_f \dot x_i|}} \exp\big\{i \left[\frac{1}{\hbar}S(x_f, x_i; E) - \mu \frac{\pi}{2}\right]\big\}
$$

For free motion, the Maslov index is 0:


$$
\mathcal{G}_{VVG}^{d = 1}(x_f, x_i, E) = \frac{1}{i \hbar \sqrt{|\dot x_f \dot x_i}} \exp\left\{i \left[\frac{1}{\hbar}S(x_f, x_i; E)\right]\right\}
$$

Now since $E = \frac{1}{2}mv^2$, we find that $v = \dot x = \sqrt{\frac{2E}{m}}$, regardless of where the particle is. 

Now as we've seen before, the classical action $S$ for this system is $p|x_f - x_i| = \hbar k |x_f - x_i|$. The only other relation we'll need along the way is $E = \frac{\hbar^2 k^2}{2m}$. So then we can plug in like so:

$$
\begin{aligned}
\mathcal{G}_{VVG}^{d = 1}(x_f, x_i, E) &= \frac{1}{i \hbar \sqrt{|\dot x ^2}} \exp\left\{i \left[\frac{1}{\hbar}\hbar k |x_f - x_i|\right]\right\} \\
&= \frac{1}{i \hbar {\dot x}} \exp\left\{i \left[ k |x_f - x_i|\right]\right\} \\
&= \frac{1}{i \hbar {\sqrt{\frac{2E}{m}}}} \exp\left\{i \left[ k |x_f - x_i|\right]\right\} \\
&= \frac{\sqrt m}{i \hbar {\sqrt{2E}}} \exp\left\{i \left[ k |x_f - x_i|\right]\right\} \\
&= \frac{\sqrt m}{i \hbar {\sqrt{2(\frac{\hbar^2 k^2}{2m})}}} \exp\left\{i \left[ k |x_f - x_i|\right]\right\} \\
&= \frac{m}{i \hbar^2 k} \exp\left\{i \left[ k |x_f - x_i|\right]\right\}, \\
\end{aligned}
$$

which matches the expression 12.78 given in the book. 

## 5: 

Compute the Van Vleck-Gutzweiller propagator (see eq 13.60) for free motion in three dimensions. Compare with the exact propagtor (12.115) and show that they are identical. 

For reference, eq 13.60:

$$
\mathcal{G}_{VVG}(r_f, r_i, E) = \sum_{l = 1}^{Q(E)} \frac{1}{i \hbar} \left(\frac{1}{2 \pi i \hbar }\right)^\frac{d-1}{2} \sqrt{|\mathcal D_1(r_F, r_i; E)|} \exp\left\{i \left[\frac{1}{\hbar}S_l(r_f, r_i, E) - \mu_l \frac{\pi}{2}\right]\right\}
$$

The first great thing we can do is drop the sum, since in free space there's only one way to get from point A to point B. We don't need to sum over multiple trajectories. 

We simply follow the example that comes right after 13.60 in the book, only using free 3d motion instead of 1d. 

Our classical action, analogous to eq 13.61, is:

$$
S(r_f, r_i; E) = \int_{r_i}^{r_f} p \,d r = p(r_f - p_i) = \sqrt{2mE}(r_f - r_i),
$$

from which we obtain the analogue to 13.62:

$$
\frac{\partial^2 S(r_f, r_i; E)}{\partial E^2} = - \frac{\sqrt{2m}}{4E^{3/2}}(r_f - r_i).
$$

Now to obtain $\mathcal{D}_R$, we can look to eq 13.21:

$$
\mathcal{D}_R = ...  = \left(\frac{m}{T}\right)^d, 
$$

which in our case is $(m/T)^3$, where that transit time from $r_i$ to $r_f$  is

$$
T = \frac{r_f - r_i}{v} = \frac{m (r_f - r_i)}{p} = \frac{m}{\sqrt{2mE}} (r_f - r_i),
$$

and thus

$$
\begin{aligned}
\mathcal{D} = -\frac{\partial^2 S(r_f, r_i; E)}{\partial E^2} \mathcal{D}_R(T) &= \frac{\sqrt{2m}}{4E^{3/2}}(r_f - r_i) \left(\frac{m}{\frac{m}{\sqrt{2mE }}(r_f - r_i)}\right )^3 \\
&= \frac{\sqrt{2m}}{4E^{3/2}}\frac{(r_f - r_i)}{(r_f - r_i)^3} \left({\sqrt{2mE}}\right)^{3} \\
&= \frac{4m^2}{4} \frac{1}{(r_f - r_i)^2} \\
&= \frac{m^2}{(r_f - r_i)^2}. \\
\end{aligned}
$$

Now, $\mathcal{D}>0 \implies \mu_l = 0$, so then  we have everything on hand we need to compute the only term in this Van-Vleck Gutzweiller propagator. 


$$
\begin{aligned}
&\mathcal{G}_{VVG}(r_f, r_i, E)  \\
&= \frac{1}{i \hbar} \left(\frac{1}{2 \pi i \hbar }\right)^\frac{d-1}{2} \sqrt{|\mathcal D_1(r_F, r_i; E)|} \exp\left\{i \left[\frac{1}{\hbar}S_l(r_f, r_i, E) - \mu_l \frac{\pi}{2}\right]\right\} \\
&= \frac{1}{i \hbar} \left(\frac{1}{2 \pi i \hbar }\right) \sqrt{\frac{m^2}{(r_f - r_i)^2}} \exp\left\{i \left[\frac{1}{\hbar}\sqrt{2mE}(r_f - r_i) \right]\right\} \\
&= -\frac{1}{2 \pi \hbar^2} \sqrt{\frac{m^2}{(r_f - r_i)^2}} \exp\left\{i \left[\frac{1}{\hbar}\sqrt{2mE}(r_f - r_i) \right]\right\} \\
&= -\frac{1}{2 \pi \hbar^2} {\frac{m}{(r_f - r_i)}} \exp\left\{i \left[\frac{1}{\hbar}\sqrt{2mE}(r_f - r_i) \right]\right\} \\
&= -\frac{1}{2 \pi \hbar^2} {\frac{m}{(r_f - r_i)}} \exp\left\{i \left[k(r_f - r_i) \right]\right\} \\
\end{aligned}
$$

We simply follow the example that comes right after 13.60 in the book, only using free 2d motion instead of 1d. 

Our classical action, analogous to eq 13.61, is:

$$
S(r_f, r_i; E) = \int_{r_i}^{r_f} p \,d r = p(r_f - p_i) = \sqrt{2mE}(r_f - r_i),
$$

from which we obtain the analogue to 13.62:

$$
\frac{\partial^2 S(r_f, r_i; E)}{\partial E^2} = - \frac{\sqrt{2m}}{4E^{3/2}}(r_f - r_i).
$$

Now to obtain $\mathcal{D}_R$, we can look to eq 13.21:

$$
\mathcal{D}_R = ...  = \left(\frac{m}{T}\right)^d, 
$$

which in our case is $(m/T)^2$, where that transit time from $r_i$ to $r_f$  is

$$
T = \frac{r_f - r_i}{v} = \frac{m (r_f - r_i)}{p} = \frac{m}{\sqrt{2mE}} (r_f - r_i),
$$

and thus

$$
\begin{aligned}
\mathcal{D} = -\frac{\partial^2 S(r_f, r_i; E)}{\partial E^2} \mathcal{D}_R(T) &= \frac{\sqrt{2m}}{4E^{3/2}}(r_f - r_i) \left(\frac{m}{\frac{m}{\sqrt{2mE }}(r_f - r_i)}\right )^2 \\

\end{aligned}
$$

Now, $\mathcal{D}>0 \implies \mu_l = 0$, so then  we have everything on hand we need to compute the only term in this Van-Vleck Gutzweiller propagator. 


$$
\begin{aligned}
&\mathcal{G}_{VVG}(r_f, r_i, E)  \\

We rewrite this in terms of $k = \sqrt{2mE}/\hbar$:

$$
\mathcal{G}_{VVG}(r_f, r_i, E) = \frac{m \cdot e^{-i(3\pi/4)}}{\hbar \sqrt{2\pi \hbar^2 k (r_f - r_i)}} \cdot e^{i k (r_f - r_i)} \\

$$
H_0^{(+)}(z) \to \sqrt{\frac{2}{\pi i z}}e^{iz}.

$$

$$

So while the prefactors of eq 1 and eq 2 might be different, they are both of the form

$$
\sqrt\frac{1}{z}e^{iz}, \quad z = k |r_f - r_i|,
