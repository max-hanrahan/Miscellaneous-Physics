## 1:

The density of states for a quantum particle of mass $m$ inside an infinite square well of width $a$ is

$$
\rho_\text{ISW} (E) = \sum_{n=1}^{\infty} \delta(E - E_n), \quad E_n = \frac{\hbar^2 \pi^2}{2 ma^2}n^2, \quad n = 1, 2...
$$

Use the energy-dependent propagator

$$
G_\text{ISW}^{(+)}(x^\prime, x; E) = \frac{m}{i \hbar^2 k} \sum_{l = -\infty}^{\infty} \left[e^{ik [x^\prime -x + 2 al ]}- e^{ik [x^\prime +x + 2 al ]}\right]
$$

of the infinite square well (see eq 12.85) to verify the general trace formula

$$
\rho(E) = - \frac{1}{\pi} \Im \text{Tr} \hat {\mathcal{G}}(E)
$$

Let's start with 14.7:

$$
\begin{aligned}
\rho(E) &= - \frac{1}{\pi} \Im \int \frac{m}{i \hbar^2 k} \sum_{l = -\infty}^{\infty} \left[e^{ik [x^\prime -x + 2 al ]}- e^{ik [x^\prime +x + 2 al ]}\right] dx \\
&= -\frac{1}{\pi} \Im \left[\frac{-im}{\hbar^2 k} \sum_{l=-\infty}^{\infty}\int_0^a e^{2ik |al|}dx - \int_0^a e^{2ik |x+ al|}dx\right]
\end{aligned}
$$

Strictly speaking, the arguments of the above exponentials only need $l$ in absolute value bars. Let's call these integrals $I_1$ and $I_2$:

$$
I_1 = \int_0^a e^{2ik |al|}dx = ae^{2ik|l|},
$$

and

$$
I_2 = \int_0^a e^{2i |x + al|}dx,
$$

which has three cases:

* $l=0$:

$$
I_2 = \int_0^a e^{2i |x|}dx = \left[\frac{e^{2ikx}}{2ik}\right]_0^a = \frac{1}{2ik} (e^{2ika} -1)
$$

* $l <0$:

$$
I_2=\int_0^a e^{2 i k(x-a l)} d x=\frac{e^{-2 i k a l}}{2 i k}\left(e^{2 i k a}-1\right)
$$

* $l >0$:

$$
I_2=\int_0^a e^{2 i k(x+a l)} d x=\frac{e^{2 i k a l}}{2 i k}\left(e^{2 i k a}-1\right)
$$

If we put the integrals back in our $\rho(E)$ formula we have:

$$
\begin{aligned}
\rho(E) &= -\frac{1}{\pi} \left(\frac{-m}{\hbar^2 k}\right) \sum_{l = -\infty}^{\infty}a e^{2ika |l|} \\
&= \frac{ma}{\pi \hbar^2 k} \sum_{l = \infty}^{\infty} e^{2ika |l|}.
\end{aligned}
$$

Now we will use the following relations:

$$
\delta[g(x)]=\sum_{n=1}^N \frac{1}{\left|g^{\prime}(x)\right|} \delta\left(x-x_n\right) \\  \delta_ \alpha(x)=\sum_{n=-\infty}^{\infty} \delta(x-n \alpha) \\ \delta_\alpha(x)= \sum_{n=-\infty}^{\infty} e^{2 \pi i m x / \alpha}
$$

Which are from eq 1.165, 1.166, and 1.191 respectively. We also donate periodicity with $\alpha$ to avoid confusion with well width $a$. 

From this it's apparent that our $\rho(E)$ looks like the periodic delta function. And thus we can recast it as:

$$
\rho(E) = \frac{ma}{\pi \hbar^2 k} \sum_{l = -\infty}^{\infty} \delta \left(\frac{ka}{\pi} - |l|\right).
$$

To apply 1.166 we need the density of states:

$$
g(E) = \frac{\sqrt{2mE}}{\hbar} \frac{a}{\pi} - |l| \implies g^\prime(E) = \frac{a}{2 \pi} \frac{2m}{\hbar} \frac{1}{\sqrt{2mE}} = \frac{ma}{\pi \hbar} \frac{1}{\sqrt{2m \frac{\hbar^2 k^2}{2m}}} = \frac{ma}{\pi \hbar^2 k}
$$

Now notice that $\delta(\frac{ka}{\pi} - |l|)$ is only non-zero at $ka = - l \pi$, or in terms of $E = \frac{\hbar^2}{2m } \left(\frac{l \pi}{a}\right)^2 = \frac{\hbar^2 \pi^2 l^2}{2ma^2}$, and we have chosen by calcuating $g(E)$ that $k = \frac{\sqrt{2mE}}{\hbar} >0$. Since $E$ depends on $l^2$, we can throw out the negative $l$ terms in the sum. And thus

$$
\boxed{\rho(E) = \frac{ma}{\pi \hbar^2 k} \sum_{l=1}^{\infty} \frac{\pi \hbar^2 k }{ma} \delta(E - E_l) = \sum_{l = 1}^{\infty} \delta(E - E_l)}
$$

## 2:

Show that

$$
\rho(E) = \text{Tr} \delta(E - \hat {\mathcal{H}})
$$

To begin, write the formula for the trace:

$$
\text{Tr} \delta(E - \hat{\mathcal{H}}) = \sum_{n=0}^{\infty} \langle n \mid \delta(E - \hat{\mathcal{H}}) \mid  n \rangle
$$

Because $\hat{\mathcal{H}} \mid n \rangle = E_n  \mid n \rangle$, and since $\mid n \rangle$ forms a complete and orthonormal set of states, we have that  $\langle n\mid  n \rangle$ = 1. We can therefore reduce to:

$$
\text{Tr} \delta(E - \hat{\mathcal{H}}) = \sum_{n=0}^{\infty} \delta(E - E_n) = \rho(E)
$$

## 3:

Derive the trace formula

$$
\rho(E) = - \frac{1}{\pi} \Im \text{Tr} \hat {\mathcal{G}}^{(+)}(E)
$$

using the integral representation

$$
\delta(E - \hat {\mathcal{H}}) = \frac{1}{2 \pi \hbar} \int_{-\infty}^{\infty} e^{i(E - \hat{\mathcal{H}})t/\hbar} dt
$$

Starting with the result from question 14.2.2:

$$
\rho(E) =\operatorname{Tr} \delta(E-\hat{\mathcal{H}})=\operatorname{Tr} \frac{1}{2 \pi \hbar} \int_{-\infty}^{+\infty} e^{i(E-\hat{\mathcal{H}}) t / \hbar} d t=\operatorname{Tr} \frac{1}{\pi n \hbar} \int_0^{+\infty} \cos \left[\frac{(E-\hat{\mathcal{H}}) t}{\hbar}\right] d t \quad \text { (even function) } 
$$

$$
=\operatorname{Tr} \operatorname{Re} \frac{1}{\pi \hbar} \int_0^{\infty} e^{i(E-\hat{\mathcal{H}}) t / \hbar} d t=\operatorname{Tr} \operatorname{Re} \frac{1}{\pi \hbar} \int_0^{\infty} e^{i E t / \hbar} e^{-i \hat{\mathcal{H}} t / \hbar} \theta(t) d t \quad\left(\int_0^{\infty} f(t) d t=\int_0^{\infty} \theta(t) f(t) d t\right) \\
$$
Then use 12.28 to cast the propagator into $\rho(E)$:
$$
\hat{\mathcal{G}}^{(+)}(t) =\theta(t) \exp \left(-\frac{i}{\hbar} \hat{\mathcal{H}} t\right).
$$

$$\rho(E)=\operatorname{Tr} \operatorname{Re} \frac{1}{\pi \hbar} \int_0^{\infty} e^{i E t / \hbar} \mathcal{G}^{(+)}(t) d t \\
$$
Then we can transform our time-dependent propagator to an energy-dependent propagator:
$$
\hat{\mathcal{G}}^{(+)}(E) =\frac{1}{i \hbar} \int_0^{+\infty} e^{i E t / \hbar} \mathcal{G}^{(+)}(t) d t
$$
so, 
$$
\rho(E)=\operatorname{Tr} \operatorname{Re} \frac{i}{\pi} \cdot \frac{1}{i \hbar} \int_0^{\infty} e^{i E t / \hbar} \mathcal{G}^{(+)}(t) d t=\operatorname{Tr} \frac{1}{\pi} \operatorname{Re} i \mathcal{G}^{(+)}(E)
$$


Since $\operatorname{Re}\{i z\}=\operatorname{Re}\{i(x+i y)\}=\operatorname{Re}\{i x-y\}=-y=-\operatorname{Im}\{z\}: \rho(E)=-\frac{1}{\pi} \operatorname{Im} \operatorname{Tr} \hat{\mathcal{G}}^{(+)}(E)$, as desired. 

## 4:

In one dimension, Weyl's formula with surface correction is

$$
N_\text{Weyl}(E) = \frac{a \sqrt{2mE}}{\pi \hbar} - \frac{1}{2}.
$$

(a) Compute the exact level density $\rho(E)$ for a quantum particle of mass $m$ restricted to, but otherwise moving freely in, the interval $[0,a]$ on the $x$ axis.

We showed in HW6, problem 1 (7.3.1), that the exact level density $\rho(E)$ of a quantum particle in an infinite square well is given by:

$$
\rho(E) = \frac{a}{2\pi \hbar} \sqrt{\frac{2m}{ E}} \left\{ 1 + 2 \sum_{m = 1}^\infty \cos\left[\frac{2 m a}{\hbar}\left(\sqrt{2m E}\right)\right]\right\}
$$


(b) Compute $N(E)$ by integrating $\rho(E)$. 

$$
\begin{aligned}
N(E) &= \int \rho(E)dE \\
& =\int \frac{a}{\pi \hbar} \sqrt{\frac{m}{2 E}}\left(1+2 \sum_{n=1}^{\infty} \cos \left(\frac{2 n a}{\hbar} \sqrt{2 m E}\right)\right) d E \\
& =\frac{a \sqrt{m}}{\pi \hbar \sqrt{2}} 2 \sqrt{E}+\frac{a \sqrt{m} 2}{\pi \hbar \sqrt{2 E}} \sum_{n=1}^{\infty}\int \left(\cos \left(\frac{2 n a}{\hbar} \sqrt{2 m E}\right) E^{-1 / 2}\right) d E+C \\
& =\frac{a \sqrt{2 m E}}{\pi \hbar}+\sum_{n=1}^{\infty} \frac{1}{n \pi} \sin \left[\frac{2 a n \sqrt{2 m E}}{\hbar}\right]+C,
\end{aligned}
$$

for integration constant $C$. 

We also know

$$
\begin{aligned}
N(E) &= \sum_{n = 1}^{\infty} \Theta(E - \frac{n^2 \pi^2 \hbar^2}{2 ma^2}) \\
&= \sum_{n=1}^{\infty} \Theta\left(\frac{\pi^2 \hbar^2}{2ma^2} - \frac{n^2 \pi^2 \hbar^2}{2ma^2}\right) \\
&= \sum_{n=1}^{\infty} \Theta\left(\frac{\pi^2 \hbar^2}{2ma^2} \left(1 - n^2\right)\right) \\
&= \Theta(0) \\
&= \frac{1}{2}
\end{aligned}
$$

So then, we can use: 

$$
N\left(\frac{\pi^2 \hbar^2}{2ma^2}\right) = \frac{1}{2}
$$

to determine our constant:

$$
\frac{a\sqrt{2m}}{\pi \hbar} \frac{\pi \hbar}{\sqrt{2m}a } + C +\sum_{n=1}^{\infty} \frac{1}{n \pi} \sin \left[\frac{2 a n \sqrt{2 m}}{\hbar} \frac{\pi}{\hbar\sqrt{2m}a}\right] = \frac{1}{2}
$$

$$
1 + C +\sum_{n=1}^{\infty} \frac{1}{n \pi} \sin \left[2n \pi \right] = \frac{1}{2}
$$

$$
1 + C  = \frac{1}{2} \implies C = -\frac{1}{2}
$$


And thus, 

$$
\boxed{N(E) = \frac{a \sqrt{2 m E}}{\pi \hbar}+\sum_{n=1}^{\infty} \frac{1}{n \pi} \sin \left[\frac{2 a n \sqrt{2 m E}}{\hbar}\right]- \frac{1}{2}}
$$

We could have also done this without integrating by simply using 7.11. 


(c) Discarding all oscillating terms in $N(E)$, obtain $N_\text{Weyl}(E)$.
All we have to do is kill the $\sin$ term(s) and then:
$$
\boxed{N(E) = \frac{a \sqrt{2 m E}}{\pi \hbar}- \frac{1}{2}}
$$


## 5:

A particle of mass $m$ moves in the potential $V(x)$. Assume that the action $S(E)$ for $V(x)$ is known, and so are the energy level $E_n$, computed according to 

$$
S(E_n) = 2 \pi \hbar \left(n + \frac{1}{2} \right), \quad n \geq 0 \text{  integer}.
$$

The potential is changed to $W(x) = \beta V(x)$, where $\beta >0$ is a constant. Denote the energy levels of the new potential $W(x)$ by $\tilde{E}_n$. 

(a) Compute an approximate, explicit formula for the energy shift $\Delta E_n = \tilde E_n - E_n$. 

We can use 
$$
S(E) = \oint p(x) dx = \oint \sqrt{2m(E-V(x))}dx
$$, 
and similarly 

$$
\tilde S(\tilde E) = \oint \sqrt{2m(\tilde E-\beta V(x))}dx = \sqrt \beta \oint \sqrt {2m(\tilde E / \beta - V(x))}dx = \sqrt \beta \oint S(\tilde E/\beta)dx.
$$

By 14.52, we can equate these actions:

$$
\tilde S (\tilde E_n) = S(E_n) = 2 \pi \hbar\left(n + \frac{1}{2}\right),
$$

and we can rewrite the action $\tilde S (\tilde E_n)$ like so:

$$
\tilde S (\tilde E_n) = \sqrt \beta  S(\tilde E /\beta) = \sqrt\beta S(E_n + \tilde E_n/ \beta - E_n),
$$

and expand:

$$
\begin{aligned}
\sqrt \beta S(\tilde E /\beta )& \approx \sqrt{\beta}\left(S\left(E_n\right)+\left(\frac{\tilde{E}}{\beta}-E_n\right) S^{\prime}\left(E_n\right)+...\right) \\
& \approx \sqrt{\beta}\left(S\left(E_n\right)+\left(\frac{\tilde{E}}{\beta}-E_n\right) T\left(E_n\right)\right).
\end{aligned}
$$

Now make the substitution: $\tilde{E_n}=E_n+\Delta E_n$

$$
\begin{aligned}
\sqrt \beta S(\tilde E /\beta )& \approx \sqrt{\beta} S\left(E_n\right)+\sqrt{\beta}\left(\frac{E_n}{\beta}+\frac{\Delta E_n}{\beta}-E_n\right) T\left(E_n\right) \\
S\left(E_n\right) & =\sqrt{\beta} S\left(E_n\right)+\sqrt{\beta} T\left(E_n\right) E_n\left(\frac{1}{\beta}-1\right)+T\left(E_n\right) \frac{\Delta E_n}{\sqrt{\beta}}.
\end{aligned}
$$

Cool, now all we have to do is solve for $\Delta E_n$:

$$
\begin{aligned}
& (1-\sqrt{\beta}) S\left(E_n\right)=\sqrt{\beta}   T\left(E_n\right) E_n\left(\frac{1}{\beta}-1\right)+T\left(E_n\right) \frac{\Delta E}{\sqrt{\beta}} \\
& (1-\sqrt{\beta}) S\left(E_n\right)-\sqrt{\beta} T\left(E_n\right) E_n\left(\frac{1}{\beta}-1\right)=\frac{\Delta E}{\sqrt{\beta}} T\left(E_n\right) \\
& \sqrt{\beta}(1-\sqrt{\beta}) \frac{S\left(E_n\right)}{T\left(E_n\right)}-\beta E_n\left(\frac{1}{\beta}-1\right)=\Delta E_n \\
& (\sqrt{\beta}-\beta) \frac{S(E_n)}{T(E_n)} - E_n\left(\frac{1}{\beta}- 1 \right)  = \Delta E_n. \\
& \therefore \boxed{\Delta E_n=E_n(\beta-1)+(\sqrt{\beta}-\beta) \frac{S\left(E_n\right)}{T\left(E_n\right)}}
\end{aligned}
$$


(b) For the special case of the harmonic oscillator, $V(x) = m \omega^2 x^2/2$, your formula for $\Delta E_n$ should be exact. Apply your formula to the harmonic oscillator and show that it is indeed exact. 

In this case, our corresponding potential $W(x) = \beta(x)$ can be expressed as:

$$
W(x) = \frac{\beta}{2} m \tilde \omega ^2 x^2 = \frac{1}{2} m (\sqrt \beta \omega)^2x^2 = \frac{1}{2} m \tilde \omega^2 x^2,
$$

where $\tilde \omega = \sqrt \beta \omega$. 

And we know the energy levels of a harmonic oscillator (4.55):

$$
E_n = \hbar \omega \left(n + \frac{1}{2}\right),
$$

$$
\tilde E_n = \hbar \tilde \omega \left(n + \frac{1}{2}\right) = \sqrt \beta \hbar \omega \left(n + \frac{1}{2}\right) = \sqrt \beta E_n,
$$

and thus

$$
\Delta E_n  = \sqrt \beta E_n - E_n = E_n (\sqrt \beta -1).
$$

Let's verify this with the formula we found in part (a):

$$
E_n=E_n(\beta-1)+(\sqrt{\beta}-\beta) \frac{S\left(E_n\right)}{T\left(E_n\right)}.
$$

We can plug in the action for one cycle of the harmonic oscillator (5.19):

$$
S(E) = 2 \pi E / \omega \implies T(E) = \frac{2\pi}{\omega} \implies \frac{S(E)}{T(E)} = E_n,
$$

so then:

$$
\begin{aligned}
\Delta E_n &= \beta E_n - E_n + E_n (\sqrt{\beta}) -\beta E_n. \\
 &= \sqrt{\beta}E_n  - E_n \\
 &= E_n(\sqrt{\beta} -1), \\
\end{aligned}
$$
which is just we expected to find. 

## 6:

Energy-scaling systems are an important class of systems for which the action $S_l(E)$ of a periodic orbit can be written as $S_l(E) = \sigma_l f(E)$, where $\sigma_l$ is a constant and $f(E)$ is known, for instance from considering the Weyl formulas. For these systems, Gutzwiller's trace formula can be written in the form

$$
\rho(E) = \sum_n \delta(E- E_n) \approx \sum_l A_lT_l(E)E^{i S_l (E)/\hbar},
$$

where the $l$-sum is over the periodic orbits of the system, $A_l$ is a constant, and 

$$
T_l(E) = \frac{\partial S_l (E)}{\partial E}.
$$

It is sometimes very difficult to find the actions of the classical periodic orbits. Suprisingly, in such cases, and for energy-scaling systems, quantum mechanics may behelpful to determine the actions of the classical periodic orbits. Assume that $\rho(E)$ is known, either determined experimentally by spectroscopy or numerically by solving the Schr√∂dinger equation. In addition, assume that $f(E)$, too, is known. Then, show that

$$
F (\sigma) = \int_0^{E_\text{max}} \rho(E) e^{-i \sigma f(E)/\hbar}dE
$$

is sharply peaked on the action constants $\sigma_l$ of the classical periodic orbits, where $f(E_\text{max})$ is large and $f(0) = 0$. Let $N(E)$ be the staircase function. Would considering

$$
G(\sigma) = \int_0^{E_\text{max}} N(E) e^{-i \sigma f(E)/\hbar}dE
$$

be a better choice?

To begin, calculate $F(\sigma)$:
$$
\begin{aligned}
& F(\sigma)=\int_0^{E_{\text {non }}} \rho(E) e^{-i \sigma f(E)} d E \\
& =\int_0^{E_{\text {max }}} \sum_l A_l \frac{\partial S_l(E)}{\partial E} e^{\frac{i}{\hbar} \left[S_l(E)-\sigma f(E)\right]} d E \\
& =\int_0^{E_{\text {max }}} \sum_l A_l \sigma_l f^{\prime}(E) e^{\frac{i}{\hbar} \left[\sigma_l-\sigma\right] f(E) } dE\\
& =\sum_l A_l \sigma_l \frac{1}{\left(\frac{i}{\hbar}\right)\left[\sigma_l-\sigma\right]}\left\{e^{\frac{i}{i}\left[\sigma_l-\sigma\right] f\left(E_{\text {max }}\right)}-1\right\} \\
& =\sum_l A_l \sigma_l \frac{e^{\frac{i}{2 \hbar}\left[\sigma_l-\sigma\right] f\left(E_{\text {max }}\right)}}{\frac{i}{2 \hbar}\left[\sigma_l-\sigma\right] \cdot 2}\left\{e ^{\frac{i}{2\hbar} [\sigma_l - \sigma]f(E_\text{max})} - e^{- \frac{i}{2\hbar} [\sigma_l - \sigma] f(E_\text{max})}\right\} \\
& =\pi \sum_l A_l \sigma_l e^{\frac{i}{2 \hbar}\left[\sigma_l-\sigma\right] f\left(E_{\text {max }}\right)} \frac{\sin \left[\frac{\left(\sigma_l-\sigma\right)}{2 \hbar} \cdot f\left(E_{\text {max }}\right)\right]}{\pi\left(\frac{\sigma_l-\sigma}{2 \hbar}\right)}
\end{aligned}
$$

Recall that one such representation of the $\delta$-function is

$$
\lim_{c \to \infty}\frac{1}{\pi} \frac{\sin (c x)}{x}
$$

by eq 1.163. 

So we can then rewrite $F(\sigma)$:

$$
\begin{aligned}
F(\sigma )& =\pi \sum_l A_l \sigma_l e^{\frac{i}{2 \hbar}\left[\sigma_l-\sigma\right] f\left(E_{\text {max }}\right)} {\delta\left(\frac{\sigma_l-\sigma}{2 \hbar}\right)} \\
&= 2 \pi \hbar \sum_l A_l \sigma_l e^{i/2\hbar [\sigma_l - \sigma] f(E_\text{max})} \delta(\sigma_l - \sigma) \\
&= 2 \pi \hbar \sum_l A_l \sigma_l  \delta(\sigma_l - \sigma).
\end{aligned}
$$

This last step is justified because the $\delta$-function will only contribute when $\sigma_l = \sigma$, and otherwise that factor will be 1. Hence, 

$$
F(\sigma) = 2 \pi \hbar \sum_l A_l \sigma_l \delta(\sigma_l - \sigma),
$$

which is sharply peaked at $\sigma = \sigma_l$. 

---

Now, what happens when we use $N(E)$ instead of $\rho(E)$?

Similar to above, we show:

$$
\begin{aligned}
& N(E)=\int_0^E \rho\left(E^{\prime}\right) d E^{\prime}=\int_0^E \sum_l A_l T_l\left(E^{\prime}\right) e^{\frac{i}{\hbar} S_l\left(E^{\prime}\right)} d E^{\prime} \\
= & \sum_l A_l \int_0^E \sigma_l f^{\prime}\left(E^{\prime}\right) e^{\frac{i}{\hbar} \sigma_l f\left(E^{\prime}\right)} d E^{\prime} \\
= & \sum_l A_l \sigma_l \frac{1}{\left(\frac{i}{\hbar}\right) \sigma_l}\left\{e^{\frac{i}{\hbar} \sigma_l f(E)}-1\right\} .
\end{aligned}
$$

Armed with this, we compute:

$$
\begin{aligned}
G (\sigma) &= \int_0^{E_\text{max}}N(E^\prime) e^{- \frac{i}{\hbar} \sigma f(E ^\prime)}d E^\prime \\
&= \int_0^{E_\text{max}} \sum_l A_l \frac{1}{\frac{i}{\hbar} \sigma_l} \left\{ e^{\frac{i}{\hbar} \sigma_l f(E^\prime)} - 1 \right\}e^{-\frac{i}{\hbar} \sigma f(E^\prime)}dE^\prime
\end{aligned}
$$

And here we are stuck. Granted, the problem tells us what $f(E^\prime)$ is, but since the factor of $f^\prime(E)$ is no longer in the integral, we cannot go further unless we know the explict formula for $f(E^\prime)$. So no, this method is not a better choice.
