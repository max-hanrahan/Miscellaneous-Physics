## 1:

The spectrum of a quantum particle of mass $m$ in the infinite square-well potential of width $a$ is

$$
E_n = \frac{\hbar^2 \pi^2}{2ma^2} (n + 1)^2, \quad n = 0, 1, 2, ... .
$$

(a) Compute the exact level density $\rho(E)$ for $E>0$. 

Let's start from 7.9:

$$
\rho(E) = g^\prime (E) \left\{ 1 + 2 \sum_{m = 1}^\infty \cos(2\pi m g(E))\right\}.
$$


Now since the energy states were specified above, we can immiediately solve for the infinite square well's $g(E)$:

$$
g(E) = \sqrt\frac{2ma^2 E_n}{\pi^2 \hbar^2} -1.
$$

Thus, 

$$
g^\prime(E) = \frac{1}{2} \sqrt{\frac{2ma^2}{\pi^2 \hbar^2 E_n}}, 
$$


and thus

$$
\rho(E) = \frac{1}{2} \sqrt{\frac{2ma^2}{\pi^2 \hbar^2 E_n}} \left\{ 1 + 2 \sum_{m = 1}^\infty \cos\left[2\pi m \left(\sqrt\frac{2ma^2E_n}{\pi^2 \hbar^2 } -1\right)\right]\right\}.
$$

We can massage this a little further, and notice the $\cos$ term has a $-2\pi m$ in the argument, which we can drop without loss of generality. 

$$
\begin{aligned}
\rho(E) = \frac{1}{2} \sqrt{\frac{2ma^2}{\pi^2 \hbar^2 E}} \left\{ 1 + 2 \sum_{m = 1}^\infty \cos\left[2\pi m \left(\sqrt\frac{2ma^2 E}{\pi^2 \hbar^2 }\right)\right]\right\} \\
\rho(E) = \frac{a}{2\pi \hbar} \sqrt{\frac{2m}{ E}} \left\{ 1 + 2 \sum_{m = 1}^\infty \cos\left[\frac{2 m a}{\hbar}\left(\sqrt{2m E}\right)\right]\right\}. \\
\end{aligned}
$$

(b) Show that $\rho_{BS}(E)$ is exact in this case. 

Let's begin with 7.20:

$$
\rho_{BS}(E) = \frac{T(E)}{2 \pi \hbar} \sum_{m = -\infty}^{\infty} e^{im \frac{S(E)}{\hbar}}
$$

The action $S(E)$ of a particle taking one round trip in an infinite square well is given by 5.16:

$$
S(E) = 2a \sqrt{2mE},
$$

and $T(E) = \frac{dS}{dE} = \frac{1}{2E^{1/2}} 2 a \sqrt{2m} = a \sqrt{\frac{2m}{E}}$.

So then before we plug into $\rho_{BS}(E)$, let's make one more modification:

$$
\rho_{BS}(E) = \frac{T(E)}{2 \pi \hbar} \sum_{m = -\infty}^{\infty} e^{im \frac{S(E)}{\hbar}} =  \frac{T(E)}{2 \pi \hbar} \left(1+ \sum_{m =1}^{\infty} \left( e^{im \frac{S(E)}{\hbar}} + e^{-im \frac{S(E)}{\hbar}}\right)\right),
$$

which we can recognize as:
$$
\rho_{BS}(E) = \frac{T(E)}{2 \pi \hbar} \sum_{m = -\infty}^{\infty} e^{im \frac{S(E)}{\hbar}} =  \frac{T(E)}{2 \pi \hbar} \left(1+ \sum_{m =1}^{\infty} 2 \cos \left[m \frac{S(E)}{\hbar} \right]\right).
$$

And from there, all we need to do is plug in $T(E)$ and $S(E)$:

$$
\begin{aligned}
\rho_{BS}(E) &=  \frac{T(E)}{2 \pi \hbar} \left(1+ \sum_{m =1}^{\infty} 2 \cos \left[m \frac{S(E)}{\hbar} \right]\right) \\
&= \frac{a }{2 \pi \hbar} \sqrt{\frac{2m}{E}} \left(1+ \sum_{m =1}^{\infty} 2 \cos \left[m \frac{2a \sqrt{2mE}}{\hbar} \right]\right), \\
\end{aligned}
$$

which, by inspection, is equivalent to what we found in part (a). As such:

$$
\boxed{\rho_{BS}(E) = \rho(E)},
$$

which is what we wanted to prove. 

## 2:

The spectrum of the harmonic oscillator is $E_n = \hbar \omega (n + 1/2)$, $n = 0, 1, 2, ...$. Therefore, one set of possible spectral separators is $E_n^< = n \hbar \omega$ and $E_n^> = (n + 1) \hbar \omega$. Notice that this is a case where $E_n^< = E_{n+1}^<$, which is not a problem at all. Show that for this set of spectral separators, together with $g(E) = [E/(\hbar \omega)] - 1/2$, equation 7.37 yields the correct harmonic oscillator energy levels $E_n$. 

Equation 7.37: 

$$
\begin{aligned}
E_n &= \left(n + \frac{1}{2}\right)E_n^> - \left(n - \frac{1}{2}\right)E_n^< - G(E_n^>) + G(E_n^<) \\
&- \frac{1}{\pi} \sum_{m=1}^{\infty} \frac{1}{m} \int_{E_n^<}^{E_n^>} \sin[2 \pi m g(E)] dE
\end{aligned}
$$


where $G(E) = \int g(E)dE$. 

Well we might as well start by evaluating $G(E)$:

$$
\begin{aligned}
G(E) = \int \left(\frac{E}{\hbar \omega} - \frac{1}{2}\right)dE \\
= \frac{E^2}{2\hbar \omega} - \frac{E}{2} + C,
\end{aligned}
$$

where we have picked up an integration constant $C$. 

So going back to 7.37, we can expand things out. But first, there's a trick we can do to eliminate the term with the integral. Let's examine that term:

$$
\frac{1}{\pi} \sum_{m=1}^{\infty} \frac{1}{m}\int_{E_n^<}^{E_n^>} \sin (2 \pi m g(E))dE
$$

The integrand is


$$
\begin{aligned}
&\sin (2 \pi m g(E)) \\
&=\sin \left[2 \pi m \left(\frac{E}{\hbar \omega} - \frac{1}{2}\right)\right] \\
&=\sin \left[\frac{2 \pi mE}{\hbar \omega}  - m \pi\right].
\end{aligned}
$$

If we define $x = \frac{2 \pi mE}{\hbar \omega}$, then

$$
\sin(x - m \pi) = (-1)^m \sin(x).
$$

Now back to the integral: 

$$
\begin{aligned}
\int_{E_n^<}^{E_n^>} \sin \left(\frac{2 \pi mE}{\hbar \omega}\right)dE  &= - \frac{\hbar \omega}{2 \pi m}\cos \left(\frac{2 \pi mE}{\hbar \omega}\right)\big |_{E_n^<}^{E_n^>} \\
&= -\frac{\hbar \omega}{2 \pi m} \left[\cos \left(\frac{2 \pi m(n+1)\hbar \omega}{\hbar \omega}\right) - \cos \left(\frac{2 \pi mn\hbar \omega}{\hbar \omega}\right)\right] \\
&= -\frac{\hbar \omega}{2 \pi m} \left[\cos \left({2 \pi m(n+1)}\right) - \cos \left({2 \pi mn}\right)\right] \\
&= -\frac{\hbar \omega}{2 \pi m}[1-1] = 0 
\end{aligned}
$$

for all $m$, regardless of this is an even or odd term. 


Thus, the integrand is zero and we can throw out the final term from 7.37. 

Let's exapand the rest of it:

$$
\begin{aligned}
E_n &= \left(n + \frac{1}{2}\right)E_n^> - \left(n - \frac{1}{2}\right)E_n^< - G(E_n^>) + G(E_n^<) 
\end{aligned}
$$

Plug in $E_n^>$ and $E_n^<$:

$$
\begin{aligned}
E_n &= \left(n + \frac{1}{2}\right)(n + 1) \hbar \omega - \left(n - \frac{1}{2}\right)n \hbar \omega- G((n + 1) \hbar \omega) + G(n \hbar \omega) 
\end{aligned}
$$

And now use the formula for $G(E)$:
$$
\begin{aligned}
E_n &= \left(n + \frac{1}{2}\right)(n + 1) \hbar \omega - \left(n - \frac{1}{2}\right)n \hbar \omega- \hbar \omega\left(\frac{((n+1)\hbar \omega)^2}{2\hbar \omega} - \frac{(n+1) \hbar \omega}{2} + C \right) + \hbar \omega\left(\frac{(n\hbar \omega)^2}{2\hbar \omega} - \frac{n \hbar \omega}{2} + C \right)
\end{aligned}
$$


This reduces to:

$$
[n^2 + 3n/2 + 1/2 - n^2  + n/2 - n^2 / 2 - n - 1/2+ n/2 + 1/2 + n^2/ 2 - n/2]\hbar \omega
$$

and fully reduces to:

$$
\boxed{[n + 1/2]\hbar \omega}
$$

for n = 0, 1, 2, ... exactly the expression for the harmonic oscillator energy levels. 

## 3:
Bohr-Sommerfeld quantization is exact for the one-dimensional hydrogen atom. Compute its exact density of states $\rho(E)$ and show that $\rho(E) = \rho_{BS}(E)$.

Let's get the exact density of states first, following the same recipe we followd in q1. The first useful thing to write down is the energy levels of the 1d hydrogen atom (6.134):

$$
E_n = -\left(\frac{ m \alpha^2}{2\hbar^2}\right) \frac{1}{n^2}, \quad n = 1, 2, 3...
$$

which we can readily invert to receive $g(E)$:

$$
|E_n| = \left(\frac{ m \alpha^2}{2\hbar^2}\right) \frac{1}{n^2} \implies g(E) = \sqrt{\left(\frac{\alpha^2 m}{2 \hbar^2 |E|}\right)}, 
$$

which we can clean up like so:

$$
g(E) =  \frac{\alpha \sqrt{2m}}{2 \hbar}|E|^{-1/2} \implies
$$

$$
g^\prime(E) = \frac{\alpha \sqrt{2m}}{4 \hbar}|E|^{-3/2}
$$

Cool. Now following our recipe, we plug these into 7.9:

$$
\begin{aligned}
\rho(E) &= g^\prime (E) \left\{ 1 + 2 \sum_{m = 1}^\infty \cos(2\pi m g(E))\right\} \\
&= \frac{\alpha \sqrt{2m}}{4 \hbar}|E|^{-3/2} \left\{ 1 + 2 \sum_{m = 1}^\infty \cos(2\pi m \frac{\alpha \sqrt{2m}}{2 \hbar}|E|^{-1/2})\right\} \\
&= \frac{\alpha \sqrt{2m}}{4 \hbar}|E|^{-3/2} \left\{ 1 + 2 \sum_{m = 1}^\infty \cos(\pi m \frac{\alpha \sqrt{2m}}{\hbar}|E|^{-1/2})\right\} \\
\end{aligned}
$$



Now let's start from 7.20 to jumpstart an expression for $\rho_{BS}(E)$:

$$
\rho_{BS}(E) = \frac{T(E)}{2 \pi \hbar} \sum_{m = -\infty}^{\infty} e^{im \frac{S(E)}{\hbar}}
$$


And we have already calculated $S(E)$ and $T(E)$ for a 1d hydrogen atom:

$$
S(E) = \alpha \pi \sqrt{\frac{2m}{|E|}}, \quad (6.118)
$$

$$
T(E) = \alpha \pi \sqrt{\frac{m}{2|E|^{3/2}}}, \quad (6.119)
$$

so we can plug these right into 7.20:

$$
\rho_{BS}(E) = \frac{\alpha \pi \sqrt{\frac{m}{2|E|^{3/2}}}}{2 \pi \hbar} \sum_{m = -\infty}^{\infty} e^{im \frac{ \alpha \pi \sqrt{{2m}/{|E|}}}{\hbar}},
$$

and similarly to how we saw it in q1, rewrite the sum of $e^{ix}$ as a sum of $\cos(x)$:

$$
\rho_{BS}(E) = \frac{\alpha \pi \sqrt{\frac{m}{2|E|^{3/2}}}}{2 \pi \hbar} \sum_{m = 1}^{\infty} 1 + 2\cos \left({m \frac{ \alpha \pi \sqrt{{2m}/{|E|}}}{\hbar}}\right).
$$

Now we simplify:

$$
\rho_{BS}(E) = \frac{\alpha  \sqrt{{2m}}}{4  \hbar} |E|^{-3/2}\sum_{m = 1}^{\infty} \left\{ 1 + 2\cos \left({m \frac{ \alpha \pi \sqrt{{2m}/{|E|}}}{\hbar}}\right)\right\}.
$$

and by inspection we clearly see

$$
\boxed{\rho(E) = \rho_{BS}(E)}
$$

for the one-dimensional hydrogen atom, as we'd hoped to show. 

## 4:

Consider the quantum graph shown in 8.3(b). 

(a) Show that the boundary conditions at each of the six internal vertices is equivalent to requiring continuity and continuity of the first derivative of the graph. 

Our conditions are:

$$
\varphi_{21}(a) = \varphi_{32}(a)
$$

$$
\varphi_{21}^\prime(a) + \varphi_{32}^\prime(a) =0
$$

Since all of the edge lengths are $a$, the wavenumber $k$ will be the same for all edges. 

$$
\varphi(x) = A e^{\pm ikx},
$$

where $A$ may be different, however we see

$$
A e^{\pm ika} = Be^{ik(0)}
$$

and we can plug in $x_{32} = a - x_{21}$:

$$
A e^{\pm ika} = Be^{ika} \implies A = B.
$$

This is equivalent to saying the wavefunction is continuous at the vertex. 

Condition 2:

$$
\pm ikAe^{\pm ika} \pm ikAe^{\pm ik(0)} = 0
$$

Now substitute in $x_{23} = a - x_{21}$:

$$
ikAe^{\pm ika} = \mp ikA e^{\pm ikA} = 0
$$


Thus, $\varphi_{21}^\prime(a) = \varphi_{32}^\prime(0)$, so the first derivative is continuous at the vertex. 

(b): Compute the spectrum. 

From what we showed in (a) the problem reduces to a 1d scattering problem, with periodic boundary conditions and a length of $6a$.

The wavefunction $\varphi(x) = e^{\pm ikx}$ is subject to $\varphi(x) = \varphi(x+L)$ with $L = 6a$. 

Then from this periodic boundary condition:

$$
e^{\pm ikx} = e^{\pm ik(x+L)}  \implies e^{\pm ikL} = 1 \implies k = \frac{2 \pi m}{L},
$$

or 

$$
\boxed{k = \frac{\pi m}{3a}, m \in \mathbb Z},
$$

which define the energy spectrum of this Benzene-esque graph.

## 5:

Compute the spectrum of the three-pronged star graph for the case $b = 2a$ and $c = a$. 

We can skip a lot of the setup by starting from equation 8.18, the general spectral equation for the three-pronged star graph:

$$
\cos(ka) \sin(kb) \sin(kc) + \sin(ka) \cos(kb) \sin(kc) + \sin(ka) \sin(kb) \cos(kc) =0
$$


Now substitute in $b = 2a$ and $c = a$:

$$
\cos(ka) \sin(2ka) \sin(ka) + \sin(ka) \cos(2ka) \sin(ka) + \sin(ka) \sin(2kb) \cos(ka) =0,
$$

and simplify:

$$
\begin{aligned}
2\cos(ka) \sin(2ka) \sin(ka) &+ \sin(ka) \cos(2ka) \sin(ka)  =0 \\
2\cos(ka) \sin(2ka) \sin(ka) &+ \sin^2(ka) \cos(2ka) =0,
\end{aligned}
$$


We can factor out $\sin(ka)$ (as long as we don't divide, we don't want singularity issues):

$$
\sin(ka)[2\cos(ka) \sin(2ka)  + \sin(ka) \cos(2ka)] =0
$$

Now we'll use $\sin(2ka) = 2\cos(ka) \sin(ka)$:

$$
\begin{aligned}
\sin(ka)[2\cos(ka) 2\cos(ka) \sin(ka)  + \sin(ka) \cos(2ka)] =0 \\
\sin(ka)[4\cos^2(ka) \sin(ka)  + \sin(ka) \cos(2ka)] =0 \\
\end{aligned}
$$

Factor out another $\sin(ka)$:

$$
\sin^2(ka)[4\cos^2(ka)   +  \cos(2ka)] =0 
$$

Finally, use  $\cos(2ka) = 2 \cos^2(ka)-1$:

$$
\begin{aligned}
\sin^2(ka)[4\cos^2(ka)   +  2 \cos^2(ka)-1] =0 \\
\sin^2(ka)[6 \cos^2(ka) - 1] =0
\end{aligned}
$$


One solution occurs when $\sin^2(ka) =0$, or $k_na = n \pi, n = 1, 2, ...$.

Let's use another index to write the solutions to setting the other factor to zero:

$$
6 \cos^2(ka) - 1 =0 \implies \cos(ka) = \pm \frac{1}{\sqrt6}
$$

Here's a nice way to write this:

$$
k_ma = m\pi + \arccos\left(\frac{1}{\sqrt6}\right), \quad m = 0, 1, 2...
$$

or

$$
k_la = l\pi - \arccos\left(\frac{1}{\sqrt6}\right), \quad l = 1, 2,...
$$


To capture all solutions, we can rewrite this as:

$$
\begin{aligned}
a k_n &= n \pi, \quad &n =1, 2, ... \\
a k_n &= n \pi + \cos^{-1}\left(\frac{1}{\sqrt6}\right), \quad &n  =0, 1, ... \\
a k_n &= n \pi - \cos^{-1}\left(\frac{1}{\sqrt6}\right), \quad &n  =1, 2, ... \\
\end{aligned}
$$


But we need to get this using _one_ index, so we can manipulate the $n$'s to agree. If we exclude $ka = 0$, then this is the correct order (but keep $n= 0, 1, 2, ...$ as our index), then we can shuffly the index like so:

$$
\boxed{
\begin{aligned}
 k_n &= a^{-1}\left(\left(\frac{n}{3}\right) \pi + \cos^{-1}\left(\frac{1}{\sqrt6}\right)\right), \quad &n =0, 3, 6,..\\
 k_n &= a^{-1}\left(\left(\frac{n+2}{3}\right)\pi - \cos^{-1}\left(\frac{1}{\sqrt6}\right)\right), \quad &n  =1, 4,7, ... \\
 k_n &= a^{-1}\left(\left(\frac{n+1}{3}\right)\right) \pi, \quad &n  =2, 5, 8... \\
\end{aligned}
}
$$


And by 8.19, the energy spectrum is given by 

$$
E_n = \frac{\hbar^2 k_n^2}{2m}, \quad n = 0, 1, 2, ...
$$

```python
# simulation for fun
import numpy as np

def ka(n):
    if n == 0 or n == 3 or n == 6:
        ka = (n)/3*np.pi + np.acos(1/np.sqrt(6))
    elif n == 1 or n == 4 or n ==7:
        ka = (n+2)/3*np.pi - np.acos(1/np.sqrt(6))
    elif n == 2 or n == 5 or n ==8:
        ka = (n+1)*np.pi/3
    else:
        return "nope"
    return(ka)
```

```python
import matplotlib.pyplot as plt
x = np.linspace(0, 6*np.pi, 500)
# y = np.sin(x)* (2*np.cos(x) * np.sin(2*x) + np.sin(x) * np.cos(2 *x))
y = np.sin(x)**2 * (6 * np.cos(x)**2 - 1)
plt.plot(x, y)
plt.axhline(0)

m = [0, 1, 2, 3, 4, 5, 6, 7, 8]

plt.plot()

for i in m:
    plt.axvline(ka(i), c= 'k', linestyle = '--')


plt.title("Spectral Equation")
plt.xlabel(r"$ka$")
plt.ylabel(r"$\sin^2(ka)[6 \cos^2(ka) - 1]$")
plt.xlim(-.1, 3 * np.pi)
```

## 6:

Use the method of scattering quantization to compute the spectrum of the two-vertex quantum graph shown in Figure 8.2. Assume Dirichlet boundary conditions at the vertices $V_j$ and $V_l$. In addition, assume that channel 1 corresponds to the quantum particle moving from $V_j$ to $V_l$ (left $\to$ right) and channel 2 corresponds to the opposite direction (right $\to$ left).

When in channel 1, a particle will accumulate phase $e^{ika}$ and reflects in the opposite direction (into ch2), so we can infer the following elements from the scattering matrix:

$$
S_{21} = -e^{-ika}, \quad S_{11} = 0.
$$


Similarly when in ch2, a particle will accumulate phase $e^{ika}$ and reflects in the opposite direction (into ch1), and thus

$$
S_{12} = -e^{-ika}, \quad S_{22} = 0.
$$

Thus:

$$
S =\begin{bmatrix} S_{11} & S_{12} \\ S_{21} & S_{22} \end{bmatrix} = \begin{bmatrix} 0 & -e^{-ika} \\ -e^{-ika} & 0 \end{bmatrix} = -e^{-ika} \begin{bmatrix} 0 &  1\\ 1 & 0. \end {bmatrix}
$$


To find the stationary states, set 

$$
S(k) \psi = 1 \psi.
$$

We know what to do from here:

$$
\det |S(k) \psi - 1| = 0 \implies \det \big|\begin{bmatrix} -1 & -e^{-ika} \\ -e^{-ika} & -1 \end{bmatrix} \big| = 0
$$

So $1 - e^{2ika} =0 \implies 2ka = 2\pi n \implies k_n = \frac{n \pi}{a}, \quad n = 1, 2, 3...$

The energy spectrum is then

$$
E_n = \frac{\pi^2 \hbar^2 n^2}{2ma^2}, \quad n =1, 2, 3...
$$

And that is the same as the infinite square well.
