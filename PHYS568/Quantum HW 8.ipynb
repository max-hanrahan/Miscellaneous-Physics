## 1:

Show that

$$
\begin{aligned}
&\lim_{\epsilon \to 0^+} \int_{-\infty}^{\infty} \frac{\exp(ia x)}{k^2 - x^2 + i\epsilon}dx \\
&=\lim_{\epsilon \to 0^+} \int_{-\infty}^{\infty} \frac{\cos(a x)}{k^2 - x^2 + i\epsilon}dx  = \frac{\pi}{i |k|} \exp(i |ak|).
\end{aligned}
$$

The first step is trivial, and  just comes from the fact that $\exp(iax) = \cos(ax) + i \sin(ax)$. Breaking up the integral like this will cause the $\sin$ integral to vanish due to odd symmetry. 

From eq 9.2, we have:

$$
\mathcal{P} \left(\frac{1}{X}\right) = \lim_{\epsilon \to 0} \frac{X}{X^2 + \epsilon^2}, 
$$

and from 9.4 we have:

$$
\lim_{\epsilon \to 0} \int_{-\infty}^\infty \frac{1}{X \pm i \epsilon} f(X) dX = \int_{-\infty}^\infty
\left[\mathcal{P} \left(\frac{1}{X}\right) \mp i \pi \delta(x)\right] f(X)dX.
$$

Now let $X = k^2 - x^2$, and by 9.5 we have:

$$
\lim_{\epsilon \to 0} \frac{1}{k^2 - x^2 \pm i \epsilon} = \mathcal{P} \left(\frac{1}{k^2 - x^2} \right) \mp i \pi \delta(k^2 - x^2).
$$

From here on out, I will only consider the case where $\pm \to +$ and $\mp \to -$, as that's the only one we're interested in solving. 

Now, by eq. 9.9:

$$
\mathcal{P} \left(\frac{1}{x^2 - k^2} \right) = \left(\frac{1}{2k}\right) \mathcal{P} \left(\frac{1}{x - k} \right) - \left(\frac{1}{2k}\right) \mathcal{P} \left(\frac{1}{x + k} \right), 
$$

so then 

$$
\mathcal{P} \left(\frac{1}{k^2 - x^2} \right)  = \frac{1}{2 |k|} \left[\mathcal{P}\left(\frac{1}{x + |k|}\right) - \mathcal{P}\left(\frac{1}{x - |k|}\right) \right]. 
$$.

We will also use the property of the delta function that

$$
\delta[g(x)] = \sum_{n=1}^{N} \frac{1}{|g^\prime(x_n)|}\delta(x - x_n)
$$

(eq. 1.1.165). 

So then for $g(x) = k^2 - x^2 = (k-x)(k+x)$, we have $g^\prime(x) = -2x$, and therefore:

$$
\delta(k^2 - x^2) = \frac{1}{|2k|} \left[\delta(x + |k|) - \delta(x - |k|)\right].
$$

We can use one last fact about the principal value:

$$
\int_{-\infty}^{\infty} e^{iax} \mathcal{P}\left(\frac{1}{x}\right)dx= i \pi \text{sgn}(a),
$$

which means, by translation, that


$$
\int_{-\infty}^{\infty} e^{iax} \mathcal{P} \left(\frac{1}{x \mp |k|}\right)dx = e^{\pm i |ka|} i \pi \text{sgn}(a), 
$$

which follows from

$$
\mathcal{P}\left(\frac{1}{x - |k|}\right) f(x) = \mathcal{P}\left(\frac{1}{x}\right) f(k + |x|).
$$

Okay, now we are ready to attack our integral. 

The first step is to rewrite it:

$$
\begin{aligned}
&\lim_{\epsilon \to 0^+} \int_{-\infty}^{\infty} \frac{\exp(ia x)}{k^2 - x^2 + i\epsilon}dx \\
&= \int_{-\infty}^{\infty} e^{iax} \left[\mathcal{P} \left(\frac{1}{k^2 - x^2}\right) + \frac{-i\pi}{2|k|} \left(\delta(x - |k|) + \delta(x + |k|)\right)\right]dx
\end{aligned}
$$

where we have expanded out the delta function. 

Let's do the principal value part first, call that $I_{PV}$:

$$
\begin{aligned}
I_{PV}&= \int_{-\infty}^{\infty} e^{iax} \mathcal{P} \left(\frac{1}{k^2 - x^2}\right) dx \\
&= \frac{1}{2|k|} \left[e^{-ia|k|} - e^{ia|k|}\right] i \pi \text{sgn} (a) \\
&= \frac{1}{2|k|} \left[-2i \sin (|ka|) \right] i \pi \text{sgn} (a) \\
&= \frac{\pi \text{sgn}(a) \sin(|ka|)}{|k|}.
\end{aligned}
$$

Now consider the delta integral, call that $I_\delta$:

$$
\begin{aligned}
I_\delta &=-\frac{-i \pi }{2 |k|} \int_{-\infty}^{\infty}e^{iax} \left[\delta(x - |k|) + \delta(x + |k|)\right] \\
&= -\frac{-i \pi }{2 |k|} [e^{ia|k|} + e^{-ia|k|}] \\
&= \frac{i \pi }{|k|} [\cos(a |k|)]. \\
\end{aligned}
$$

Excellent. Now we combine:

$$
I = I_{PV} + I_\delta = \frac{\pi \text{sgn}(a) \sin(|ka|)}{|k|} + \frac{i \pi }{|k|} [\cos(a |k|)] = \frac{i \pi }{|k|}e^{i|ka|},
$$


and we arrive at the desired result. 

## 2:

Show that
$$
I = \int \int_D \cdots \int \exp\left(- \frac{1}{2}x^T M x \right) d^n x = \left[\frac{(2\pi)^n}{\det(M)}\right]^{1/2},
$$

where $D$ is the entire $n$-dimensional coordinate space and $M$ is a constant, real, symmetric matrix whose eigenvalues are all positive. 

The multi-dimensional stationary phase approximation is _almost_ useful here (eq 9.80), but we will re-derive it with a complex function so we don't have to worry about phases that appear. 


$$
\begin{aligned}
& \iint_D \cdots \int \exp \left(-\frac{1}{2} \vec{x}^{T} M \vec{x}\right) d^n \vec{x}
= \iint_D \cdots \int \exp \left(-\frac{1}{2} \vec{x}^{T} \Omega^{T} \Lambda \Omega \vec{x}\right) d^n \vec{x}, \quad M=\Omega^{T} \Lambda \Omega, \quad \Lambda=\operatorname{diag}\left(\lambda_1, \cdots, \lambda_n\right) \\
& \quad= \iint_D \cdots \int \exp \left(-\frac{1}{2}(\Omega \vec{x})^{T} \Lambda(\Omega \vec{x})\right) d^n \vec{x} \\
& \quad= \iint_D \cdots \int \exp \left(-\frac{1}{2} \vec{\xi}^{T} \Lambda \vec{\xi}\right) d^n \vec{\xi}, \text{ where } \vec{\xi} = \Omega \vec{x} \\
& \quad= \iint_D \cdots \int \exp \left[-\frac{1}{2}
\begin{pmatrix}
\xi_1 & \xi_2 & \cdots & \xi_n
\end{pmatrix}
\begin{pmatrix}
\lambda_1 & 0 & \cdots & 0 \\
0 & \lambda_2 & \cdots & 0 \\
\cdots & \cdots & \ddots & \cdots \\
0 & 0 & \cdots & \lambda_n
\end{pmatrix}
\begin{pmatrix}
\xi_1 \\
\xi_2 \\
\vdots \\
\xi_n
\end{pmatrix}
\right] d^n \vec{\xi} \\
& \quad= \iint_D \cdots \int \exp \left[-\frac{1}{2} \left(\lambda_1 \xi_1^2 + \lambda_2 \xi_2^2 + \cdots + \lambda_n \xi_n^2\right) \right] d^n \vec{\xi} \\
& \quad= \prod_{j=1}^n \int_{-\infty}^{+\infty} \exp \left(-\frac{1}{2} \lambda_j \xi_j^2\right) d\xi_j
\end{aligned}
$$

We recognize this as a product of Gaussian integrals: $\int_{-\infty}^{+\infty} e^{-a x^2} d x=\sqrt{\frac{\pi}{a}}$, and thus we reduce to 

$$
\prod_{j=1}^n \sqrt{\frac{2\pi}{\lambda_j}}, 
$$

and we can write the product of eigenvalues as

$$
\lambda_1 \cdot \lambda_2 \cdot ... \lambda_n = \det \Lambda,
$$

which is also equivalent to 

$$
\det (\Omega ^T \Lambda \Omega ) = \det M
$$

by the determinant theorem. 

We can therefore substitute the above in our product of Gaussian integrals, and arrive at

$$
\boxed{I = \int \int_D \cdots \int \exp\left(- \frac{1}{2}x^T M x \right) d^n x = \left[\frac{(2\pi)^n}{\det(M)}\right]^{1/2},}
$$

as desired. 

## 3:

Show that the exact, time-dependent, half-space propagator $\mathcal G_\text{HS}^{(+)} (x_f t_f, x_i t_i)$ for a quantum particle of mass $m$, moving freely in $x >0$, with a totally reflecting wall at $x=0$ is given by

$$
\mathcal G_\text{HS}^{(+)} (x^\prime \, t^\prime, x \, t) = \sqrt{\frac{m}{2 \pi i \hbar (t^\prime - t)}} \left\{\exp \left[i \frac{m(x^\prime - x)^2}{2 \hbar (t^\prime -t)}\right] - \exp \left[i \frac{m(x^\prime + x)^2}{2 \hbar (t^\prime -t)}\right]\right\}
$$

For a free particle of mass $m$ in 1d, eq 12.51 gives us the following propagator for $t^\prime >t$:

$$
\mathcal G_\text{free}^{(+)} (x^\prime \, t^\prime \, , x\, t) = \sqrt{\frac{m}{2 \pi i \hbar (t^\prime - t)}} \exp \left[i \frac{m(x^\prime - x)^2}{2 \hbar (t^\prime -t)}\right]
$$

For a wall at $x = 0$, the wavefunction must vanish at the wall: $\psi(x = 0, t) =0$ (Dirichlet boundary conditions). 

That means the proagator must vanish when either $x = 0$ or $x^\prime =0$. 

To enforce this we can use the method of images: we treat $x^\prime$ as the "source" coordinate, reflect it accross the wall to $-x^\prime$, and subtract its contribution. This will ensure the particle never crosses $x =0$. 

$$
\begin{aligned}
\mathcal G_\text{HS}^{(+)} (x^\prime \, t^\prime, x \, t) &= \sqrt{\frac{m}{2 \pi i \hbar (t^\prime - t)}} \left\{\exp \left[i \frac{m(x^\prime - x)^2}{2 \hbar (t^\prime -t)}\right] - \exp \left[i \frac{m(-x^\prime - x)^2}{2 \hbar (t^\prime -t)}\right]\right\} \\
&=\boxed{ \sqrt{\frac{m}{2 \pi i \hbar (t^\prime - t)}} \left\{\exp \left[i \frac{m(x^\prime - x)^2}{2 \hbar (t^\prime -t)}\right] - \exp \left[i \frac{m(x^\prime + x)^2}{2 \hbar (t^\prime -t)}\right]\right\}}, \\
\end{aligned}
$$

and we have our desired result. 

## 4:

Keeping only the first two terms of the Born series in 

$$
\hat {\mathcal G}(E) = \hat G(E) + \hat G(E) \hat V \hat G(E) + \hat G(E) \hat V \hat G(E) \hat V \hat G(E) + ...,
$$

obtain an approximate result for the energy-dependent propagator $\mathcal G^{(+)} (x^\prime, x; E)$ of the Hamiltonian:

$$
\mathcal{\hat H} = -\frac{\hbar^2}{2m} \frac{d^2}{dx^2} + \begin{cases}
0, \quad \text{for } |x| > a, \\
-V_0, \quad \text{for } |x| <a,
\end{cases}
$$

where $V_0>0$ and $a >0$. Focus on the case where $x <-a$, $x^\prime >a$, and $E>0$.

To begin, we'll use the result of 12.4.1, namely that the free particle propagator of the free-particle Hamiltonian is explicitly

$$
\mathcal{G}_0^{(+)}(x^\prime, x; E) = \frac{m}{i \hbar^2 k}e^{ik |x^\prime - x|}, 
$$

where I'm using the 0-index to indicate that is the first term in this truncated Born series. The second term is given by

$$
\mathcal{G}_1^{(+)}(x^\prime, x; E) =\int _{-a}^a dx_1 \mathcal{G}^+_0(x^\prime, x_1) V(x_1)\mathcal{G}^+_0(x_1, x).
$$

Notice that we have changed the bounds from $(-\infty, \infty) \to (-a, a)$ at the question's suggestion. 

So then we can be more explicit:

$$
\begin{aligned}
\mathcal{G}_1^{(+)}(x^\prime, x; E) &= -V_0 \int_{-a}^{a}dx_1 \left[\frac{m}{i \hbar^2 k}e^{ik |x^\prime - x_1|}\right] \left[\frac{m}{i \hbar^2 k}e^{ik |x_1 - x|}\right] \\
&= -V_0 \left(\frac{m}{i \hbar^2 k}\right)^2e^{ik(x^\prime - x)}  \int_{-a}^{a} dx_1 \\
&= -V_0 \left(\frac{m}{i \hbar^2 k}\right)^2e^{ik(x^\prime - x)}  (2a) \\
&= V_0 \left(\frac{2m^2a}{\hbar^4 k^2}\right)e^{ik(x^\prime - x)}  \\
\end{aligned}
$$

So then the propagator is approximated like so:

$$
\begin{aligned}
\mathcal G^{(+)} (x^\prime, x; E) &\approx  \frac{m}{i \hbar^2 k}e^{ik |x^\prime - x|} + V_0 \left(\frac{2m^2a}{\hbar^4 k^2}\right)e^{ik(x^\prime - x)} \\
&= \frac{m}{i\hbar^2k}e^{ik(x^\prime - x)} \left[1 + \frac{2i mV_0a}{\hbar^2 k}\right], 
\end{aligned}
$$

and we're done. 

## 5:

In the text we showed that only the direct, classical trajectory connecting $(x_i, t_i)$ with $(x_f, t_f)$ contributes to the free-motion propagor; all other Feynman paths interfere away. Another example is the harmonic oscillator (see Eq 12.54). Keeping only the direct, classical trajectory $P$ connecting $(x_i, t_i)$ with $(x_f, t_f)$, show that the exponential factor of $\mathcal G ^{(+)}(P)$ in Eq 12.112 is the same as the exponential factor of the exact propagator $\mathcal G_\text{OSC} (x_f t_f, x_i t_i)$ stated in Equation 12.55. 


For reference, eq 12.55:

$$
\begin{aligned}
& \mathcal{G}_{\mathrm{osc}}^{(+)}\left(x^{\prime} t^{\prime}, x t\right)=\sqrt{\frac{m \omega}{2 \pi i \hbar \sin \left[\omega\left(t^{\prime}-t\right)\right]}} \\
& \quad \exp \left\{\left(\frac{i m \omega}{2 \hbar}\right) \frac{\left(x^{\prime 2}+x^2\right) \cos \left[\omega\left(t^{\prime}-t\right)\right]-2 x x^{\prime}}{\sin \left[\omega\left(t^{\prime}-t\right)\right]}\right\} .
\end{aligned}
$$

and eq 12.112:

$$
\mathcal{G}^{(+)}(P) =\mathcal{N}(P) \exp[i S[P]/\hbar]
$$

where $\mathcal{N}$ is a normalization constant and $S(P)$ is the classical action. 

Thus, it suffices to show that the two exponential terms are equivalent when we solve for the classical action of a harmonic oscillator. 

Well, we know the Lagrangian for the H.O.:

$$
L = T - V = \frac{1}{2}m\dot{x}^2 - \frac{1}{2}m\omega^2 x^2, 
$$

and we know the trajectory:

$$
x(t) = A\cos(\omega t) + B\sin(\omega t).
$$

Unfortunately we haven't specified boundary conditions that will get super clean expressions for these coefficients, but they aren't too bad if we don't want to lose generality:

$$
x_i = A\cos(\omega t_i) + B\sin(\omega t_i)
$$


$$
x_f = A\cos(\omega t_f) + B\sin(\omega t_f)
$$

are our boundary conditions, and thus we can solve this system of equations to receive

$$
A = \frac{x_i \sin(\omega t_f) - x_f \sin(\omega t_i)}{\sin[\omega(t_f - t_i)]}
$$

$$
B = \frac{-x_i \cos(\omega t_f) + x_f \cos(\omega t_i)}{\sin[\omega(t_f - t_i)]}, 
$$

and we can rewrite our trajectory as

$$
x(t) = x_i \frac{\sin[\omega(t_f - t)]}{\sin[\omega(t_f - t_i)]} + x_f \frac{\sin[\omega(t - t_i)]}{\sin[\omega(t_f - t_i)]}, 
$$

from which we find

$$
\dot x(t) = \frac{\omega \left(x_{f} \cos{\left(\omega \left(t - t_{i}\right) \right)} - x_{i} \cos{\left(\omega \left(t - t_{f}\right) \right)}\right)}{\sin{\left(\omega \left(t_{f} - t_{i}\right) \right)}}, 
$$

Plugging into Lagrangian,

$$
L=\frac{1}{2} m \dot{x}^2-\frac{1}{a} m \omega^2 x^2
$$

this picks up an $\omega^2$ term \& flips signs

$$
L=\frac{1}{2} m \omega^2\left[A^2\left(\sin ^2 \omega t+\cos ^2 \omega t\right)+B^2\left(\cos ^2 \omega t-\sin ^2 \omega t\right)-4 A B \sin \omega t \omega \cos k t\right]
$$

Useful integrals to have on hand:

$$
\begin{aligned}
& \int\left(\sin ^2 \cdot \cos ^2\right)dt=-\int \cos 2 \omega tdt=-\left.\frac{1}{2 \omega} \sin 2 \omega t\right|_0 ^t \\
& \int\left(\cos ^2-\sin ^2\right)dt=\int \cos 2 \omega tdt=\left.\frac{1}{2 \omega} \sin 2 \omega t\right|_0 ^t \\
& \int \sin \omega t \cos \omega t dt=\frac{1}{2} \int \sin 2 \omega tdt=\left.\frac{1}{2} \sin 2 \omega t\right|_0 ^t
\end{aligned}
$$

The action is now
$$
S=\frac{m \omega}{4}\left[\left(B^2-A^2\right)\left[\sin 2 \omega t_f-\sin 2 \omega t_0\right]-2 A B\left[\cos 2 \omega t_f-\cos 2 \omega t_0\right]\right]
$$

From there we can make use of a few more things:

$$
\begin{aligned}
B^2-A^2&=\frac{x_f^2\left(\cos ^2 \omega t_0 \sin ^2 \omega t_0\right)+x_0^2\left(\cos ^2 \omega t_f-\sin ^2 \omega t_f\right)}{\sin ^2\left(\omega\left(t_f-t_0\right)\right)} \\
&=\frac{\cos 2 \omega t_0\left(x_f^2-x_0^2\right)-x_0 x_f \cos \left(\omega\left(t_0-t_f\right)\right)}{\sin ^2\left(\omega\left(t_f-t_0\right)\right)}
\end{aligned}
$$

$$
\begin{aligned}
A B&= \frac{x_0 x_f\left(\sin \omega t_f \cos \omega t_0-\cos \omega t_f \sin \omega t_0\right)+\left(x_0^2+x_f^2\right)(\cos \omega t_f \sin \omega t_f)}{\sin ^2\left(\omega\left(t_f-t_0\right)\right)} \\
&= \frac{x_0 x_f\left( \sin (\omega (t_f - t_0))\right)+\left(x_0^2+x_f^2\right)(\cos \omega t_f \sin \omega t_f)}{\sin ^2\left(\omega\left(t_f-t_0\right)\right)} \\
\end{aligned}
$$

We plug these values back into our classical action and receive:

$$
S(E) = \frac{m \omega}{2 \sin \omega T} \left[ (x_f^2 + x_i^2) \cos \omega T - 2 x_f x_i \right], \quad T = t_f - t_i
$$

Okay, back to the problem at hand. From here, we may plug the classical action back into 12.112:

$$
\mathcal{G}^{(+)}(P) =\mathcal{N}(P) \exp\left[\frac{i \frac{m \omega}{2 \sin \omega T} \left[ (x_f^2 + x_i^2) \cos \omega T - 2 x_f x_i \right]}{\hbar}\right], 
$$

and a little simplification ($t_f \to t^\prime, t_i \to t$ ) yields

$$
\mathcal{G}^{(+)}(P) =\mathcal{N}(P) \exp \left\{\left(\frac{i m \omega}{2 \hbar}\right) \frac{\left(x^{\prime 2}+x^2\right) \cos \left[\omega\left(t^{\prime}-t\right)\right]-2 x x^{\prime}}{\sin \left[\omega\left(t^{\prime}-t\right)\right]}\right\},
$$

and therefore the two exponential factors are identical. 

## 6:

Verify that the energy-dependent propagator $\mathcal G_0^{(+)}(r^\prime, r; E)$ for free motion in three dimensions is

$$
\mathcal{G}_0^{(+)}(r^\prime,r; E) = - \frac{m}{2 \pi \hbar^2 |r^\prime - r|}e^{ik |r^\prime -r|},
$$

where $r^T = (x,y,z)$.

The process will be similar to what's on p 341, only with 3d instead of 1d. 

The analogue of 12.66 becomes

$$
\hat {\mathcal H}_0 (\hat r) = - \frac{\hbar^2}{2m} \nabla^2.
$$

Meanwhile 12.67 is the same as it was before, but note that $\hat k$ is now a vector (unless otherwise unmarked in this problem). 

It is still the case that, for each $k$-vector, the energy spectrum is 

$$
E_{\vec k} = \frac{\hbar^2 k^2}{2m},
$$

but in three dimeninsions, 

$$
\langle \vec r \mid \vec k \rangle = \psi_{kx}(x) \psi_{ky}(y) \psi_{kz}(z)
$$

where $\psi_{kx} = \frac{1}{\sqrt{2\pi}}e^{ik_x x}$.

As such, we can plug the wavefunction into $\langle \vec r \mid \vec k \rangle$ and we will see

$$
\langle \vec r \mid \vec k \rangle = \frac{1}{(2 \pi)^{3/2}} e^{i \vec{k} \cdot \vec{r}}
$$


must be true for the wavefunctions to be orthonormal and complete. 

Now, the free propagator is given by:

$$
\mathcal{G}_0^{(+)}(\vec r^\prime, \vec r; E) = \int_{-\infty}^{\infty} \frac{\psi_{k^\prime}(\vec r ^\prime) \psi^{\ast}_{k^\prime}(\vec r )}{(E - E_{\vec k ^\prime}) + i \epsilon} d^3 k^\prime,
$$

which is the three dimensional analogoue of 12.72. 

Let's first write the spherial coordinate element in $k$-space:

$$
d^3 k^\prime = k^2 \sin \theta d k d \theta d \phi.
$$

Our free propagator then becomes:

$$
\begin{aligned}
\mathcal{G}_0^{(+)}(\vec r^\prime, \vec r; E) &= \frac{1}{(2 \pi)^3} \int_{-\infty}^\infty \frac{e^{i \vec k \cdot (\vec r^\prime - \vec r)}}{-E_{\vec k} + E + i \epsilon}d^3 k^{\prime} \\
&= \frac{1}{(2\pi)^3} \int_0^\infty k^2 dk \int_0^\pi \sin \theta d \theta \int_0^{2 \pi} \frac{e^{i \vec k^\prime \cdot (\vec r^\prime - \vec r)}}{-E_{\vec k} + E + i \epsilon} \\
&= \frac{1}{(2\pi)^3} \int_0^\infty k^2 dk \int_0^\pi \sin \theta d \theta \int_0^{2 \pi} \frac{e^{i \vec k^\prime R \cos \theta}}{-E_{\vec k} + E + i \epsilon} \\
&= \frac{1}{(2\pi)^3} \int_0^\infty k^2 dk \int_0^\pi \sin \theta d \theta (2\pi)\frac{e^{i \vec k^\prime R \cos \theta}}{-E_{\vec k} + E + i \epsilon} \\
&= \frac{1}{(2\pi)^2} \int_0^\infty k^2 dk \int_0^\pi \sin \theta d \theta \frac{e^{i \vec k^\prime R \cos \theta}}{-E_{\vec k} + E + i \epsilon} \\
\end{aligned}
$$


The $d\theta$ integral is done in Griffith's Quantum,  eq 10.59:

$$
\int_0^\pi e^{isr \cos \theta} \sin \theta d \theta = 2 \sin(sr)/sr,
$$

which simplifies our calculation:

$$
\begin{aligned}
\mathcal{G}_0^{(+)}(\vec r^\prime, \vec r; E) &= \frac{1}{(2\pi)^2} \int_0^\infty k^2 dk \int_0^\pi \sin \theta d \theta \frac{e^{i \vec k^\prime R \cos \theta}}{-E_{\vec k} + E + i \epsilon} \\
&= \frac{1}{(2\pi)^2} \int_0^\infty dk  \frac{ k^2}{-E_{\vec k} + E + i \epsilon} \left(\frac{2 \sin (k^\prime R)}{k^\prime R}\right) \\
\end{aligned}
$$

We'll also make the following substitutions:

$$
\bar k = \sqrt{k^2 + i \bar \epsilon}
$$

where 

$$
\bar \epsilon = \frac{2m\epsilon}{\hbar^2}.
$$

Now our integral reduces to:

$$
\frac{2}{(2\pi)^2 R} \cdot -\frac{-2m }{\hbar^2} \int_0^{\infty} \frac{k^\prime \sin(k^\prime R)}{k^{\prime 2} - \bar k ^2}dk
$$

Now let's break the $\sin$ apart into its complex exponential form.

$$
\frac{2}{(2\pi)^2 R} \left[\frac{1}{2} \cdot \frac{1}{2i} \int_{-\infty}^\infty \frac{k^\prime e^{ik^\prime R}}{(k^\prime - \bar k) (k^\prime + \bar k)}dk - \frac{1}{2} \cdot \frac{1}{2i} \int_{-\infty}^\infty \frac{k^\prime e^{-ik^\prime R}}{(k^\prime - \bar k) (k^\prime + \bar k)}dk \right],
$$

where I have snuck in a factor of two and extended the bounds from $-\infty \to +\infty$ as the integral is symmetric. 

Following Griffiths 10.60-10.64, this reduces to

$$
\begin{aligned}
\mathcal{G}_0^{(+)}(\vec r^\prime, \vec r; E) &= \frac{2}{(2\pi)^2 R}  \left[\frac{1}{2} \frac{i}{2i}\pi e^{i \bar k R} + \frac{1}{2} \frac{i}{2i}\pi e^{i \bar k R}\right] \\
&= \frac{-m}{2 \pi R \hbar^2} \left[\frac{e^{i \hat k R}}{2} + \frac{e^{i \hat k R}}{2}\right] \\
&= \frac{-m}{2 \pi |\vec r^\prime - \vec r| \hbar^2} \left[\frac{e^{i \hat k |\vec r^\prime - \vec r|}}{2} + \frac{e^{i \hat k |\vec r^\prime - \vec r|}}{2}\right] \\
&= \frac{-m}{2 \pi |\vec r^\prime - \vec r| \hbar^2}  \left[{e^{i \hat k |\vec r^\prime - \vec r|}} \right].\\
\end{aligned}
$$

Lastly, take $\epsilon \to 0$ such that $\bar k \to k$. 

This finally grants us our desired result:

$$
