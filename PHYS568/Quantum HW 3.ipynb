A quantum particle is confined to the interval $[-a, a]$. It is described by the time-dependent wave function
$$
\begin{aligned}
\psi(x, t) & =\frac{1}{\sqrt{2 a}}\left\{\cos \left(\frac{\pi}{2 a} x\right) \exp \left[-i\left(\frac{\hbar \pi^2}{8 m a^2}\right) t\right]\right. \\
& \left.-\sin \left(\frac{\pi}{a} x\right) \exp \left[-i\left(\frac{\hbar \pi^2}{2 m a^2}\right) t\right]\right\}
\end{aligned}
$$
(a) Show that $\psi(x, t)$ is properly normalized, i.e., show that
$$
\int_{-a}^a \rho(x, t) d x=\int_{-a}^a|\psi(x, t)|^2 d x=1
$$
for all $t$. 


First, we'll need the complex conjugate on hand:

$$
 \psi^\ast(x, t) =\frac{1}{\sqrt{2 a}}\left\{\cos \left(\frac{\pi}{2 a} x\right) \exp \left[i\left(\frac{\hbar \pi^2}{8 m a^2}\right) t\right] -\sin \left(\frac{\pi}{a} x\right) \exp \left[i\left(\frac{\hbar \pi^2}{2 m a^2}\right) t\right]\right\}
$$

For simplicity, let's define $E_1 = \frac{\hbar^2 \pi^2}{8 m a^2}$ and $E_2 = \frac{\hbar^2 \pi^2}{2 m a^2}$

Then we have:

$$
\begin{aligned}
\psi ^\ast \psi(x,t) &= \frac{1}{2a} \left\{\cos^2 \left(\frac{\pi}{2a}x\right) + \sin^2 \left(\frac{\pi}{a}x \right) - \cos \left(\frac{\pi}{2a}x\right) \sin \left(\frac{\pi}{a}x \right) \left(e^{i (E_1 - E_2)t/\hbar} + e^{-i(E_1 - E_2)t/\hbar}\right)\right\} \\
&= \frac{1}{2a} \left\{\cos^2 \left(\frac{\pi}{2a}x\right) + \sin^2 \left(\frac{\pi}{a}x \right) - 2\cos \left(\frac{\pi}{2a}x\right) \sin \left(\frac{\pi}{a}x \right)  \cos {\left(\frac{E_1 - E_2}{\hbar} t\right)}\right\} \\
\end{aligned}
$$

Notice that over this integral, the term containing both $\cos$ and $\sin$ will disappear due to orthogonality. 

Thus, we are left with:

$$
\int_{-a}^a|\psi(x, t)|^2 d x= \int_{-a}^{a}\frac{1}{2a} \left\{\cos^2 \left(\frac{\pi}{2a} x\right) + \sin^2 \left(\frac{\pi}{a}x \right)\right\} dx
$$

Which we can break into two integrals:

$$
\begin{aligned}
\int_{-a}^{a} \cos^2 \left(\frac{\pi}{2a}x\right) dx &= a \\
\int_{-a}^{a} \sin^2 \left(\frac{\pi}{a}x\right)dx &= a
\end{aligned}
$$

So then the square of the probability amplitude becomes:

$$
|\psi(x,t)|^2 =\frac{1}{2a} (a + a) = \boxed{1}
$$

and hence, the wave function is demonstrably normalized.

$$

We now integrate over the left side of our probability space. 

$$
P_\text{left}(t) = \frac{1}{2a} \int_{-a}^0\left\{\cos^2 \left(\frac{\pi}{2a}x\right) + \sin^2 \left(\frac{\pi}{a}x \right) - 2\cos \left(\frac{\pi}{2a}x\right) \sin \left(\frac{\pi}{a}x \right)  \cos {\left(\frac{E_1 - E_2}{\hbar} t\right)}\right\}dx \\

Let's compute this integral one piece at a time:

And now we must show that the time-integral from $0$ to $\tau$ of the probability current is equal to this value. 

Let's use eq. 4.32 as the definition of our probability current. 

$$
J(x,t) = \frac{\hbar}{2 m i} \left[\psi^\ast (x, t) \frac{\partial \psi(x, t)}{\partial x} - \psi(x, t) \frac{\partial \psi^\ast(x, t)}{\partial x}\right]
$$

So then, 

$$
J(0,t) = \frac{\hbar}{2 m i} \left[\psi^\ast (0, t) \frac{\partial \psi(0, t)}{\partial x} - \psi(0, t) \frac{\partial \psi^\ast(0, t)}{\partial x}\right]
$$

Let's break this into pieces:

$$
\psi(0, t) = \frac{e^{\frac{i \pi ^2 \hbar^2t}{8 a^2 m}}}{\sqrt{2} \sqrt{a}}
$$

$$
\psi^\ast(0, t) = \frac{e^{-\frac{i \pi ^2 \hbar^2 t}{8 a^2 m}}}{\sqrt{2} \sqrt{a}}
$$

$$
\frac{\partial \psi(0, t)}{\partial x} = -\frac{\pi  e^{\frac{i \pi ^2 \hbar^2t}{2 a^2 m}}}{\sqrt{2} a^{3/2}}
$$

$$
\frac{\partial \psi^\ast(0, t)}{\partial x} = -\frac{\pi  e^{-\frac{i \pi ^2 \hbar^2t}{2 a^2 m}}}{\sqrt{2} a^{3/2}}
$$

Now we assemple the pieces:

$$
\begin{aligned}
J[0, t] &= \frac{h}{2mi} \left[ \left(\frac{e^{-\frac{i \pi ^2 \hbar^2 t}{8 a^2 m}}}{\sqrt{2} \sqrt{a}}\right) \left(-\frac{\pi  e^{\frac{i \pi ^2 \hbar^2t}{2 a^2 m}}}{\sqrt{2} a^{3/2}}\right) - \left(\frac{e^{\frac{i \pi ^2 \hbar^2 t}{8 a^2 m}}}{\sqrt{2} \sqrt{a}}\right) \left(-\frac{\pi  e^{-\frac{i \pi ^2 \hbar^2t}{2 a^2 m}}}{\sqrt{2} a^{3/2}}\right)\right] \\
&= \frac{h}{2mi}\left(\frac{\pi  e^{-\frac{3 i \pi ^2 \hbar^2 t}{8 a^2 m}} \left(-1+e^{\frac{3 i \pi ^2 \hbar^2 t}{4 a^2 m}}\right)}{2 a^2}\right) \\
&= \frac{h}{2mi} \left(\frac{i \pi  \sin \left(\frac{3 \pi ^2 \hbar^2 t}{8 a^2 m}\right)}{a^2}\right) \\
&= \frac{h}{2m} \left(\frac{ \pi  \sin \left(\frac{3 \pi ^2 \hbar^2 t}{8 a^2 m}\right)}{a^2}\right) \\
\end{aligned}
$$

At last, we can integrate with respect to $t$ from $0$ to $\tau$:

$$
\begin{aligned}
\int_0^\tau J(0, t)dt &= \int_0^\tau \frac{h}{2m} \left(\frac{ \pi  \sin \left(\frac{3 \pi ^2 \hbar^2 t}{8 a^2 m}\right)}{a^2}dt\right) \\
&= -\left(\frac{h \pi}{2ma^2}\right) \left(\frac{8 a^2 m}{3 \pi^2 \hbar^2}\right) \cos\left(\frac{3 \pi^2 \hbar^2 t}{8 a^2m}\right) \big |_0^{\tau} \\
&= - \frac{4}{3 \pi} \left[\cos\left(\frac{3 \pi^2 \hbar^2 \tau}{8 a^2m}\right) - \cos \left(0\right) \right] \\
&= -\frac{4}{3 \pi} \left[\cos\left(\frac{3 \pi}{2}\right) - \cos (0)\right] \\
&= \frac{4}{3 \pi}
\end{aligned}
$$

as desired.

## 2:

Compute the probability of finding a quantum particle in the ground state of the harmonic oscillator $(n =0)$ in the classically forbidden regions beyond the two classical turning points. 

Let's start from the wavefunction for a particle in the harmonic oscillator's ground state:

$$
\psi_0 = \frac{1}{\sqrt[4]{b^2 \pi}} \exp\left(-\frac{x^2}{2b}\right)
$$

and ground state energy $E_0$ is $\frac{1}{2} \hbar \omega$. 

The turning points are when $V(x) = E$, which we can solve:

$$
\frac{1}{2}m \dot \omega^2 x^2 = \frac{1}{2} \hbar \omega \implies x_\pm = \pm \sqrt\frac{\hbar}{m \omega} = \pm b
$$

So then the probability of finding the particle in the classically forbidden regions is really the union of the probabilities of finding it beyond the left and right turning points:

$$
P_\text{out} = \int_{-\infty}^{x_-}|\psi_0|^2 dx + \int_{x_+}^{\infty}|\psi_0|^2 dx
$$

These integrals will not be that bad, especially because $\psi$ is real and not complex. In addition, we can take advantage of the symmetric wave function and obtain:

$$
\begin{aligned}
P_\text{out} &= 2 \int_{x_+}^{\infty} |\psi_0(x)|^2 dx \\
&= 2 \left(\frac{m \omega}{\pi \hbar} \right)^{1/2} \int_{x_+}^{\infty} e^{-m \omega x^2 / h}dx
\end{aligned}
$$

This is a Guassian integral which can be expressed in terms of the complementary error function $\text{erfc}$:

$$
\begin{aligned}
P_\text{out} &=2 \left(\frac{m \omega}{\pi \hbar} \right)^{1/2} \frac{1}{2} \sqrt{\frac{\pi \hbar}{m \omega}}\text{erfc} \left(\sqrt{\frac{m \omega}{\hbar}}x_+\right) \\
&= \text{erfc} \left(\sqrt{\frac{m \omega}{\hbar}}x_+\right) \\
&= \text{erfc} \left(1\right) \\
&= \boxed{\approx  0.1573}\\
\end{aligned}
$$

## 5:

Show that a quantum particle of mass $m$ moving in the attractive $\delta$-function potential $V(x) = -V_0 \delta(x), V_0 > 0$, has exactly one bound state. Compute its wave function and its binding energy.  


Bound states occur when $E<0$. To the left of the spike in the potential, this means that:

$$
-\frac{\hbar^2}{2m} \psi^{\prime \prime}(x) = E \psi(x)
$$

If we define $\kappa^2 = - \frac{2mE}{\hbar^2}$, then we have a simple differential equation:

$$
\psi^{\prime \prime}(x) = \kappa^2  \psi(x)
$$

Note that this means $\kappa^2 >0$, since we are in an $E <0$ region. 

The solution to this differential equation is 

$$
\psi_L = A e^{\kappa x} + Be^{- \kappa x}
$$

Similarly, for the region to the right of the potential's zenith, we can solve for the right half of the wavefunction $\psi_R$:

$$
\psi_R = Ce^{\kappa x} + De^{-\kappa x}
$$

We can kill both the $B$ and $C$ terms, because if they were alive they would blow up to infinity as $x \to \infty$. 

Our general wavefunction is then

$$
\psi(x) = \begin{cases}
Ae^{\kappa x}, \quad &x \leq 0 \\
De^{-\kappa x}, \quad &x \geq 0
\end{cases}
$$

Further, this must obey Dirichlet boundary conditions: namely, that the value of $\psi(x)$ is continuous at $x = 0$. 

Thus, $A = D$, and our wavefunction becomes:

$$
\psi(x) = \begin{cases}
Ae^{\kappa x}, \quad &x \leq 0 \\
Ae^{-\kappa x}, \quad &x \geq 0.
\end{cases}
$$

But we can do better. The $\delta$-function has to introduce a discontinuity in the derivative of the wave function at $x = 0$. We can differentiate each side:

$$
\begin{aligned}
&\psi_L^\prime(x) = \kappa Ae^{\kappa x} \\
&\psi_R^\prime(x) = -\kappa Ae^{-\kappa x} \\
\end{aligned}
$$

This is helpful because we can differentate this across an infinitesimal interval:

$$
\lim_{\epsilon \to 0} \left[\frac{d \psi}{dx} |_{x = \epsilon} - \frac{d \psi}{dx}|_{x = -\epsilon}\right]
$$

and we can plug in our left and right $\psi$'s to find:

$$
\psi_{R}^{\prime }(0) - \psi_L^{\prime}(0) = -A\kappa -A\kappa= -2A \kappa
$$

But of course we can also calculate this difference in  derivatives by using the Schrodinger equation, and then 

$$
\lim_{\epsilon \to 0} \left[\frac{d \psi}{dx} |_{x = \epsilon} - \frac{d \psi}{dx}|_{x = -\epsilon}\right] = - \frac{2mV_0}{\hbar^2} \psi(0)
$$


And we know $\psi(0) = A$ based on our Dirichlet boundary conditions. We equate these two and simplify to receive:

$$
\kappa = \frac{mV_0}{\hbar^2}
$$

This is the only allowed $\kappa$. And we can transform this back into its corresponding energy:

$$
\boxed{E = -\frac{mV_0^2}{2 \hbar^2}.}
$$

This is the only allowed energy level, which corresponds to the ground state of this potential. 

Of coutse, the wave function must be normalized. We can take advatage of even symmetry:

$$
\begin{aligned}
\int_{-\infty}^{\infty} |\psi(x)|^2dx &= 2\int_{0}^{\infty} |\psi(x)|^2dx \\
&= 2\int_{0}^{\infty}A^2 e^{-2 \kappa x}dx \\
&= 2A^2 \left(-\frac{1}{2 \kappa}\right) \left(e^{-\infty} - e^0\right) \\
&= \frac{A^2}{\kappa}
\end{aligned}
$$

And thus, $A = \sqrt{\kappa} = \sqrt{\frac{mV_0}{\hbar^2}}$, and the normalized wave function is

$$
\boxed{\psi(x) = \sqrt{\frac{m V_0}{\hbar^2}} e^{-\frac{mV_0}{\hbar^2} |x|}.}
$$
