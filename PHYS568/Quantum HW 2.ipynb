## 1:

 A particle of mass m moves without friction on the triangular track shown in Fig 3.2. Starting at point A with coordinates $(x = -a, y = 0)$ with initial speed $v_0$, the particle may reach
point C with coordinates $(x = a, y = 0)$ in two different ways:
1. directly via the straight section AC or 
2. via the V-shaped part of the track, ABC, where the particle is under the influence of a gravitational field with gravitational acceleration $g$ in the negative $y$ direction. The distance from A to C on AC is $2a$; the legs AB and CB of the V-shaped section of the track form an angle $\phi < \pi/2$ with the $x$ axis; the coordinates of B are
$(x = 0, y = -b)$, where $b = a\tan (\varphi)$.


(a) Compute the time $T_h$ it takes the particle to move from A to C on the horizontal part AC of the track. 

For uniform motion on a linear track,

$$
d = v_0 t
$$

Therefore, the time it takes to move a distance $2a$ at speed $v_0$ is 

$$
\boxed{T_H = \frac{2a}{v_0}}
$$

(b) Compute the time $T_V(\varphi)$ it takes the particle to reach point C via point B on the V-shaped part of the track. Express your result in terms of $T_H$, $\varphi$, and the dimensionless parameter $\beta = ga/v_0^2$. 

We can calculate the length $L_{AB} = L_{BC}$ using the pythagorean formula. Since it is given that $b = a\tan(\varphi)$:

$$
L_{AB} = L_{BC} = \sqrt{a^2 + b^2} = \sqrt{a^2 + a\tan (\varphi)}
$$

We can then take advantage of trig relations:

$$
L_{AB} = L_{BC} = a\sqrt{1 + \tan ^2 \varphi} = \frac{a}{\cos \varphi}
$$

We know from kinematics that, for constant acceleration $a$ (distinct from other uses of $a$ in this problem), we can produce distance $x$ using $x = v_0 t + \frac{1}{2} a t^2$. \
The vertical component of acceleration in this case is $g \sin \varphi$, and thus our kinematic equation becomes:

$$
\frac{a}{\cos \varphi} = v_0 t_1 + \frac{1}{2} g \sin (\varphi) t_1^2
$$

We can then solve this quadratic equation for $t_1$, the time it takes to move from point $A$ to point $B$. This is effectively our initial velocity

$$
- \frac{a}{\cos \varphi} + v_0 t_1 + \frac{1}{2} g \sin (\varphi) t_1^2 = 0
$$

This has the following solution:

$$
t_1 = \frac{-(v_0) \pm \sqrt{(v_0)^2 + 2( g a \tan (\varphi) ) }}{ g \sin (\varphi) }
$$

Let's also calculate the final velocity at point $B$, $v_B$:

$$
v_B = \sqrt{v_0^2 + 2 ga \tan \varphi}
$$

This is useful in another kinematic equation, $v_f = v_i + at$ for  acceleration $a$ (not to be confused with our distance $a$). Then we can use this to determine what our initial velocity was:

$$
v_0 = \sqrt{v_0^2 + 2 ga \tan \varphi} - g \sin \varphi t_2
$$

Where $t_2$ is the time interval to traverse $BC$. 

We can thus solve for $t_2$:
$$
t_2 = -\frac{v_0- \sqrt{v_0^2 + 2 ga \tan \varphi}}{g \sin \varphi } 
$$


And by construction, $T_H$ is the sum of $t_1$ and $t_2$, meaning:

$$
T_V = t_1 + t_2 = \frac{-(v_0) + \sqrt{(v_0)^2 + 2( g a \tan (\varphi) ) }}{ g \sin (\varphi) }-\frac{v_0- \sqrt{v_0^2 + 2 ga \tan \varphi}}{g \sin \varphi } 
$$

Where we have taken the positive root of $t_1$, cause we don't want a negative time interval. Reduce:

$$
T_V = \frac{-2v_0 + 2\sqrt{(v_0)^2 + 2( g a \tan (\varphi) ) } }{ g \sin (\varphi) }
$$

Almost there. This is in terms of $\varphi$, but certainly not in terms of $\beta$ or $T_H$. Let's fix that by using $g = v_0^2 \beta/a$:

$$
\begin{aligned}
T_V &= \frac{-2v_0 + 2\sqrt{(v_0)^2 + 2( (v_0^2 \beta/a) a \tan (\varphi) ) } }{ (v_0^2 \beta/a) \sin (\varphi) } \\
&= \frac{-2v_0 + 2\sqrt{v_0^2 + 2 (v_0^2 \beta) \tan (\varphi)  } }{ (v_0^2 \beta/a) \sin (\varphi) } \\
&= \frac{-2v_0 + 2v_0\sqrt{ 1+ 2 \beta \tan (\varphi)  } }{ (v_0^2 \beta/a) \sin (\varphi) } \\
&= \frac{-2 + 2\sqrt{ 1+ 2 \beta \tan (\varphi)  } }{ (v_0 \beta/a) \sin (\varphi) } \\
&=\frac{2a}{\beta v_0 \sin \varphi} \left[-1 + \sqrt{1 + 2 \beta \tan \varphi}\right]
\end{aligned}
$$

Now just recognize that $T_H = \frac{2a}{v_0}$, and, voilà!

$$
\boxed{T_V = \frac{T_H}{\beta  \sin \varphi} \left[-1 + \sqrt{1 + 2 \beta \tan \varphi}\right]}
$$
(c) Expanding $T_V(\varphi)$ to first order in $\varphi$, show that for small $\varphi$ it is faster to take the V-shaped route, i.e., for small $\varphi$, $T_V(\varphi) < T_H$:

First, let's have a few useful small angle approximations ready-at-hand:

$$
\cos \varphi \approx 1 - \frac{\varphi^2}{2} \\
\sin \varphi \approx \tan \phi \approx \phi
$$

It is also convenient that the series expansion of $\sqrt{1 + 2 \beta \tan \varphi} \approx \sqrt{1 + 2 \beta  \varphi}$ can be written as $1 + \beta \phi - \frac{(\beta \varphi)^2}{2}$ for $\phi$ close to zero. 

Putting this altogether, we can make the following approximation:

$$
\begin{aligned}
T_V &\approx \frac{T_H}{\beta  \varphi} \left[-1 + \sqrt{1 + 2 \beta  \varphi}\right] \\
&\boxed{T_V \approx {T_H} \left[1 - \frac{\beta \varphi}{2}\right]}
\end{aligned}
$$

And since $\beta = ga/v_0^2$ must be positive, as well as $\varphi$, this means that $T_V < T_H$ in the small $\varphi$ limit. It is advantageous to take the V-shaped path under gravity. 

(d) Considering the limit $\varphi \to \pi/2$, show that (i) for sufficienty large $\varphi$ it is slower to take the V-shaped route and (ii) there exists an angle $\varphi^\ast (\beta) \neq 0$ at which $T_V(\varphi^\ast (\beta)) = T_H$. 

Let's look back at our expression for $T_H$, the boxed expression at the end of part (c). 

As $\varphi \to \pi/2$, $\tan \varphi \to \infty$, which means our $T_V$ blows up. This makes sense: if the V-shape track becomes an infinite square well, the particle will classically never leave the well, and spend infinite time down there. 

But we can use this fact with what we found in part $c$: namely that for small $\varphi$, $T_V < T_H$ and now we see the opposite is true for large $\varphi$. Therefore by the intermediate value theorem there must exist a $\varphi^\ast$ such that $T_V(\varphi^\ast) = T_H$. 

Setting $T_V = T_H$ gives us:

$$
\begin{aligned}
T_H &= \frac{T_H}{\beta  \sin \varphi ^ \ast} \left[-1 + \sqrt{1 + 2 \beta \tan \varphi^\ast}\right] \\
1 &= \frac{1}{\beta  \sin \varphi ^ \ast} \left[-1 + \sqrt{1 + 2 \beta \tan \varphi ^ \ast}\right]
\end{aligned}
$$

Which we coould theoretically solve for $\phi^\ast$, but it would be far more tractable to solve for it numerically, as we'll end up with a cubic equation regardless.

## 2:

As shown in Figure 3.4, a pearl of mass $m$ and energy $E< 2mgb$ slides without friction on a wire bent into the shape of a cycloid parametrized according to 


$$
x(\tau) = b[\tau + \sin (\tau)], \quad y(\tau) = 2b \sin^2 \left(\frac{\tau}{2}\right),
$$

where $b$ is a fixed scaling parameter with dimensions of length and $\tau$ is a parameter with $-\pi \leq \tau \leq \pi$. Under the influence of a gravitational field with gravitational accelration $ g$ in the negatie $y$ direction, the pearl executes an oscillatory motion with amplitude $x_\text{max} = a$ and oscillation period $T$. Compute $T$. You will find that $T$ is independent of the oscillation amplitude $a$...


Since we have $x$ and $y$ displacement vectors, we should get their time derivatives, as they will be useful to construct the Hamiltonian:

$$
\begin{aligned}
\dot x(\tau) = b[1 + \cos (\tau)] \dot \tau, \quad \dot y(\tau) &= 2b\dot \tau \sin \left(\frac{\tau}{2}\right) \cos \left(\frac{\tau}{2}\right), \\
&= b \dot \tau \sin \tau
\end{aligned}
$$

We'll soon need to call upon $\dot x^2 + \dot y^2$ for our kinetic energy contribution, might as well do that out now:

$$
\begin{aligned}
\dot x^2 + \dot y^2 &= \left(b^2 \dot \tau^2 +  2b^2 \dot \tau^2 \cos \tau +  b^2 \dot \tau^2 \cos^2 \tau\right) +  \left(b^2 \dot \tau^2 \sin^2 \tau\right) \\
&= b^2 \dot \tau^2 +  2b^2 \dot \tau^2 \cos \tau +  b^2 \dot \tau^2 \\
&= 2b^2 \dot \tau^2 +  2b^2 \dot \tau^2 \cos \tau \\
&= 2b^2 \dot \tau^2 (1+ \cos \tau). \\
\end{aligned}
$$

Good. Now here we can evaluate the total energy $E$:

$$
E = \frac{1}{2} mv^2 + mgy
$$

Plug in our expression for $y(\tau)$:

$$
E = \frac{1}{2} mv^2 + mg( 2b \sin^2 \left(\frac{\tau}{2}\right))
$$

Now solve for $v^2$:

$$
\begin{aligned}
  \frac{1}{2} mv^2 &= E-mg( 2b \sin^2 \left(\frac{\tau}{2}\right))  \\
v^2 &= \frac{2E}{m}-g( 4b \sin^2 \left(\frac{\tau}{2}\right))
\end{aligned}
$$

Now set that equal to our $\dot x^2 + \dot y^2$ expression from earlier:


$$
2b^2 \dot \tau^2 (1+ \cos \tau) = \frac{2E}{m}-g( 4b \sin^2 \left(\frac{\tau}{2}\right))
$$

This is enough to solve for $\dot \tau^2$:

$$
 \dot \tau^2 = \frac{\frac{2E}{m}-g( 4b \sin^2 \left(\frac{\tau}{2}\right))}{2b^2 (1 + \cos \tau)}
$$
And now transform back into $\frac{d\tau}{dt}$, and take the square root:
$$
 \frac{d\tau}{dt} = \sqrt{\frac{\frac{2E}{m}-g( 4b \sin^2 \left(\frac{\tau}{2}\right))}{2b^2 (1 + \cos \tau)}}
$$

One way of extracting the period is to integrate $dt$ with respect to $\tau$ with an upper bound of $\tau_\text{max}$, this is equal to 1/4 of the period. Mathematically:

$$
\begin{aligned}
\tau/4 = \int_0^{t/4} dt&= \int_0^{\tau_\text{max}} \sqrt{\frac{2b^2 (1 + \cos \tau)}{\frac{2E}{m}-g( 4b \sin^2 \left(\frac{\tau}{2}\right))}}d\tau \\
&= \int_0^{\tau_\text{max}} \left[\frac{b^2 (1 + \cos \tau)}{\frac{E}{m}-g( 2b \sin^2 \left(\frac{\tau}{2}\right))}\right]^{1/2}d\tau \\
\end{aligned}
$$

It remains to determine what $\tau_\text{max}$ is so we can evaluate the integral. This is a stationary point, and therefore there is zero kinetic energy. Thus, 

$$
E = mgy_\text{max} = 2mgb \sin^2 (\tau_\text{max}/2)
$$

and so, 

$$
\frac{E}{mb} = 2g \sin^2 (\tau_\text{max}/2)
$$

which we can plug back in to our integral:

$$
\tau/4 = \int_0^{\tau_\text{max}} \left[\frac{b (1 + \cos \tau)}{(2g \sin^2 (\tau_\text{max}/2))-g( 2 \sin^2 \left(\frac{\tau}{2}\right))}\right]^{1/2} d\tau
$$

Had to use mathematica for this, but it can be done:

$$
\tau/4 = \pi \tau_\text{max} \sqrt{\frac{b \cot (\tau_\text{max}/2)}{g \tau_\text{max}}} \sqrt\frac{\tan(\tau_\text{max}/2)}{\tau_\text{max}}
$$

The tangent and cotangent cancel, and so do the $\tau_\text{max}$'s. Would you look at that, we never even needed to compute $\tau_\text{max}$, really! 

At last:
$$
\begin{aligned}
\frac{\tau}{4} = \pi \sqrt\frac{b}{g} \\
\boxed{{\tau} = 4\pi \sqrt\frac{b}{g} }
\end{aligned}
$$

## 3: 
A particle of mass $m$ starts at $x_0 <0$ at $t = 0$ and flies to the right with energy $E >0$ in the inverted harmonic oscillator potential $V(x) = -m\omega^2 x^2/2$. Compute its trajectory $x(t)$. 

First, let's write the Hamiltonian for this system:

$$
\begin{aligned}
\mathcal{H} &= \frac{p^2}{2m} + V(x) \\
&= \frac{1}{2}m\dot x^2 + V(x) \\
&=  \frac{1}{2}m\dot x^2 -\frac{1}{2}m\omega^2 x^2
\end{aligned}
$$

Since we know this is a Newtonian system, we can cut to the chase a little bit by using the Newtonian equation of motion:

$$
\begin{aligned}
m \ddot x &= -\frac{dV}{dx}  \\
m \ddot x &= - m \omega^2 x
\end{aligned}
$$

Solving for $\ddot x$ yields a differential equation:

$$
\ddot x = -\omega^2 x
$$

To which we know the answer conforms to the following ansatz:

$$
x(t) = Ae^{\omega t} + B e^{-\omega t}
$$

We can then plug in our initial/boundary conditions to determine $A$ and $B$. 

At time $t = 0$:

$$
x(0) = A+B = x_0 <0
$$

Good to know. Meanwhile the initial condition on our velocity is:

$$
\dot x (0)= A \omega - B \omega = v_o
$$

which we obtained by differentiating our ansatz for $x(t)$. 

We then have a system of equations

$$
A+ B = x_0, \quad A- B = \frac{v_0}{\omega}
$$

which is easy enough to solve. Adding and subtracting yields:

$$
A = \frac{1}{2} \left(x_0 + \frac{v_0}{\omega}\right), \quad B = \frac{1}{2} \left(x_0 - \frac{v_0}{\omega}\right)
$$

Which we can then plug in to our ansatz:

$$
x(t) = \left(\frac{1}{2} \left(x_0 + \frac{v_0}{\omega}\right)\right)e^{\omega t} + \left(\frac{1}{2} \left(x_0 - \frac{v_0}{\omega}\right)\right) e^{-\omega t}
$$

Great. Now let's expand this out:

$$
x(t) = \frac{x_0}{2}e^{\omega t} + \frac{v_0}{2\omega}e^{\omega t} + \frac{x_0}{2}e^{-\omega t} - \frac{v_0}{2\omega}e^{-\omega t}
$$

We can therefore write this in terms of $\cosh$ and $\sinh$:

$$
x(t) =x_0 \cosh (\omega t ) + \frac{v_0}{\omega } \sinh (\omega t)
$$

Then we can use the Hamiltonian to recast this in terms of energy:

$$
x(t) =x_0 \cosh (\omega t ) + \frac{1}{\omega } \left(\sqrt {\frac{2E}{m} + \omega^2 x_0^2}\right) \sinh (\omega t)
$$

We can say we're done here, but we can also take advantage of some hyperbolic trig identities to create a more compact form.

For instance, we can take advantage of:

$$
\sinh(A + B) = \sinh A \cosh B + \cosh A \sinh B.
$$


In doing so we receive:

$$
\boxed{x(t) = \sqrt \frac{2E}{m \omega^2} \sinh \left[\omega t + \sinh^{-1} \left(\sqrt \frac{m \omega^2}{2E}x_0\right)\right]}
$$


## 4:
According to the equation $\phi(\tau) = \sqrt{2\epsilon} \sin(\tau)$, the dimensionless angular frequency of the gravitational pendulum in the small-angle approximation is $\omega =1$. This corresponds to a round-trip time of $\tau_0 = 2\pi/\omega = 2\pi$. Show that in the limit of small energies, i.e. $\epsilon \to 0$, this result is consistent with $\tau_0 = 4 \sqrt{\frac{\ell}{g}} \mathbf{K} \left(\sqrt{\frac{E}{2mg\ell}}\right)$

Going back a bit in the chain, we saw in eq 3.59 that a more general form for the round trip time of the gravitational pendulum was $\tau_0 = 4  \mathbf K \left(\sqrt{\frac{\epsilon}{2}}\right)$. Let's be explicit about the form of this elliptic integral:

$$

Now we shall consider the $\epsilon \to 0$ limit. The $\sin^2$ will go to zero, so

\end{aligned}
$$

Bingo! This is exactly what we hoped to prove. 

## 5:

Compute $T(E)$ and $S(E)$ for the quartic oscillator potential 

$$
V(x) = V_0 (\alpha x)^4
$$

where $V_0$ and $\alpha$ are positive constants with dimensions of energy and inverse length, respectively. 

To begin, we specify the Hamiltonian of the system:

$$
\mathcal H = \frac{p^2}{2m} +V_0 (\alpha x)^4
$$

At the turning points, the particle has zero velocity and therefore no kinetic energy. At this point, the total energy is just the potential energy:

$$
E = V_0 (\alpha x_\text{turn})^4
$$

which means the two real turning points of this potential are:

$$
x_\text{turn} = ± \frac{1}{\alpha}\left(\frac{E}{ V_0}\right)^{1/4}
$$

As was shown in problem 2, we can take the positive solution $x_\text{max}$ as our upper bound and compute the period as four times the integral of the traversal time. To be very explicit:

$$
\dot x = \frac{dx}{dt} \implies dt = \frac{dx}{\dot x}
$$

and then:

$$
T = 4 \int_0^{x_\text{max}} \frac{dx}{\dot x}
$$

Now the from the Hamiltonian, we can solve for $\dot x$:

$$
\dot x = \sqrt{\frac{2}{m}\left[E - V_0 (\alpha x)^4\right]}
$$

And substitute this into the period equation:
$$
\begin{aligned}
T &= 4 \int_0^{x_\text{max}} \frac{dx}{\sqrt{\frac{2}{m}\left[E - V_0 (\alpha x)^4\right]}} \\
&= 4 \sqrt{\frac{m}{2}} \int_0^{x_\text{max}} \frac{dx}{\sqrt{E - V_0 (\alpha x)^4}} \\
\end{aligned}
$$

Now let's purge the $x$ from here, by making $x_\text{max}$ dimensionless. Define $u = x/x_\text{max} \implies x = u x_\text{max}$ and thus $du  = dx/x_\text{max} \implies dx = x_\text{max} du$. 

This u-substition transforms our integral:


$$
\begin{aligned}
T(E) &= 4 \sqrt{\frac{m}{2}} \int_0^1 \frac{x_\text{max} du}{\sqrt{E - V_0 (\alpha u x_\text{max})^4}} \\
&= \frac{4}{\alpha}\left(\frac{E}{ V_0}\right)^{1/4}\sqrt{\frac{m}{2}} \int_0^1 \frac{du}{\sqrt{E - V_0 \left(\alpha u \frac{1}{\alpha}\left(\frac{E}{ V_0}\right)^{1/4}\right)^4}} \\
&= \frac{4}{\alpha}\left(\frac{E}{ V_0}\right)^{1/4}\sqrt{\frac{m}{2}} \int_0^1 \frac{du}{\sqrt{E - V_0 \left(\alpha u \frac{1}{\alpha}\left(\frac{E}{ V_0}\right)^{1/4}\right)^4}} \\
&=   \frac{4}{\alpha}\left(\frac{E}{ V_0}\right)^{1/4}\sqrt{\frac{m}{2E}} \int_0^1 \frac{du}{\sqrt{1 - u^4}}
\end{aligned} \\
$$


Last time, we showed this integral was equal to

$$
\int_0^1\frac{dx}{\sqrt{1- x^4}} = \frac{1}{\sqrt 2} \mathbf{K} \left(\frac{1}{\sqrt 2}\right)
$$

And thus, 

$$
T(E) = \frac{4}{\alpha}\left(\frac{E}{ V_0}\right)^{1/4}\sqrt{\frac{m}{2E}}  \frac{1}{\sqrt 2} \mathbf{K} \left(\frac{1}{\sqrt 2}\right)
$$

Which we can simplify to become:

$$
\boxed{T(E) = \frac{2}{\alpha}\left(\frac{E}{ V_0}\right)^{1/4}\sqrt{\frac{m}{E}}  \mathbf{K} \left(\frac{1}{\sqrt 2}\right)}
$$



Now let's move on to the accumulated action $S(E)$. Since this is an oscillator potential, this is a periodic orbit, so we can compute the following closed integral:

$$
S(E) = \oint p dx
$$

But it is far easier to use

$$
\begin{aligned}
S(E) = \int_0^E T(E)dE &= \int_0^E \frac{2}{\alpha}\left(\frac{E^\prime}{ V_0}\right)^{1/4}\sqrt{\frac{m}{E^\prime}}  \mathbf{K} \left(\frac{1}{\sqrt 2}\right)d E^\prime \\
&=  \frac{2 \sqrt m}{\alpha}\left(\frac{1}{ V_0}\right)^{1/4}  \mathbf{K} \left(\frac{1}{\sqrt 2}\right) \int_0^E \frac{1}{\left(E^\prime\right)^{1/4} }d E^\prime \\
&= \frac{2 \sqrt m}{\alpha}\left(\frac{1}{ V_0}\right)^{1/4}  \mathbf{K} \left(\frac{1}{\sqrt 2}\right) \frac{4}{3} E^{3/4} \\
&\boxed{= \frac{8 \sqrt m}{3\alpha}\left(\frac{1}{ V_0}\right)^{1/4}  \mathbf{K} \left(\frac{1}{\sqrt 2}\right) E^{3/4}} \\
\end{aligned} 
$$

## 6:

A classical particle of mass $m$ moves in the globally confining potential 

$$
V(x) = V_0 \cosh(\alpha x), \quad V_0 > 0, \quad \alpha >0
$$

Show that a single traversal of the primitive periodic at energy $E$ takes the time

$$
T(E) = \frac{4\sqrt{2m}}{\alpha \sqrt{E+V_0}} K\left(\sqrt{\frac{E-V_0}{E+V_0}}\right)
$$

and accumulates the action

$$
S(E) = \frac{8 \sqrt{2m}}{\alpha} \sqrt{E + V_0} \left[\mathbf{K}\left(\sqrt\frac{E-V_0}{E+V_0}\right) - \mathbf{E} \left(\sqrt{\frac{E-V_0}{E+V_0}}\right) \right]
$$

First, as we've seen before, find the stationary points from the Hamiltonian:

$$
E = V_0 \cosh (\alpha x_\text{max}) \implies x_\text{max} = \frac{1}{\alpha} \cosh^{-1} \left(\frac{E}{V_0}\right)
$$

We'll also need $\dot x$, which we can get from energy conservation:

$$
E = \frac{1}{2} m \dot x^2 + V_0 \cosh (\alpha x_\text{max}) \implies \dot x = \pm \sqrt{\frac{2}{m} [E - V_0 \cosh (\alpha x)]}
$$

We'll take the positive solution and, as we've done before, integrate from 0 to $x_\text{max}$ to arrive at one-fourth the period:

$$
\begin{aligned}
T(E) &= 4 \int_0^{x_\text{max}} \frac{dx}{\dot x} \\
&= 4 \sqrt\frac{m}{2}\int_0^{x_\text{max}} \frac{dx}{\sqrt{E - V_0 \cosh (\alpha x)}}
\end{aligned}
$$

Now we have to bring elliptical integrals into the picture. 
The definion of $\mathbf {K}(k)$ is, from 2.37e:
$$
\mathbf{K}(k) = \int_0^{\pi/2} \frac{d\theta}{\sqrt{1 - k^2 \sin^2 \theta}}
$$

Let's make the following trigonometric substitution:

$$
\sin \theta = \sqrt{\frac{E - V_0 \cosh (\alpha x)}{E-V_0}}
$$

Our integral then becomes


$$
T(E) = 4 \sqrt\frac{m}{2} \int_{\pi/2}^0 \frac{1}{\sin \theta} \frac{1}{\sqrt{E-V_0}} d(\sin \theta)
$$

We need to get the dummy variable to be more explict as well:

$$
d (\sin \theta) = \cos \theta \frac{d \theta}{dx} = - \frac{aV_0 \sinh (\alpha x)}{2(E-V_0) \sin \theta}.
$$

So then our integral becomes:

$$
T(E) = 4 \sqrt\frac{m}{2} \int_{\pi/2}^0 \frac{1}{\sin \theta} \frac{1}{\sqrt{E-V_0}} \left(- \frac{aV_0 \sinh (\alpha x)}{2(E-V_0) \sin \theta}\right)
$$

From here we can use $\sinh x = \sqrt{\cosh ^2 x -1}$:

$$
\sinh (\alpha x) = \sqrt{\cosh^2 (\alpha x) - 1}
$$

And $\sin^2 \theta = \frac{E - V_0 \cosh (\alpha x)}{E-V_0} \implies \frac{E-(E-V_0)\sin^2 \theta}{V_0} = {  \cosh (\alpha x)}$

From all of this, we can put these together and receive:

$$
\sinh (\alpha x) = \frac{\sqrt{E-V_0} \sqrt{E+V_0}}{V_0}\cos(\theta) \sqrt{1 - \frac{E-V_0}{E+ V_0} \sin^2 \theta}
$$

That goes back into the integral and we go back to $dx$, aiming to transform it to $d\theta$:


$$
dx = - \frac{2}{\alpha} \sqrt\frac{E-V_0}{E+V_0} \frac{\sin \theta}{\sqrt{1 - \frac{E-V_0}{E+V_0}} \sin^2 \theta} d\theta
$$

The integral can thus be massaged:


$$
T(E) = \frac{4 \sqrt{2m}}{\alpha \sqrt{E+ V_0}} \int_0^{\pi/2} \frac{d \theta}{\sqrt{1- \frac{E-V_0}{E+V_0} \sin^2 \theta}}
$$

And at last, we can employ the definition of $\mathbf{K}(k)$, where $k = \sqrt\frac{E-V_0}{E+V_0}$:

$$
\begin{aligned}
T(E) &= \frac{4 \sqrt{2m}}{\alpha \sqrt{E+ V_0}} \mathbf{K}(k) \\
&\boxed{= \frac{4 \sqrt{2m}}{\alpha \sqrt{E+ V_0}} \mathbf{K}\left(\sqrt\frac{E-V_0}{E+V_0}\right)} \\
\end{aligned}
$$

Which is what we wished to show. 

To go from period $T(E)$ to accumulated action $S(E)$, the easiest way is to recognize the following relationship:

$$
T(E) = \frac{dS(E)}{dE}
$$

It'll therefore be useful to have derivatives of the elliptical integrals on hand, given by eq. 2.38:

$$
\begin{aligned}
\frac{d \mathbf{K}(k)}{dk} &= \frac{1}{k} \left[\frac{\mathbf{E}(k)}{1 - k^2} - \mathbf{K}(k)\right] \\
\frac{d \mathbf{E}(k)}{dk}&= \frac{1}{k} [\mathbf{E} (k) - \mathbf{K}(k)]
\end{aligned}
$$

As before, we will take $k = \sqrt\frac{E-V_0}{E+V_0}$.

Then the action we want to show can be written more compactly:

$$
S(E) = \frac{8\sqrt{2m}}{\alpha} \sqrt{E+ V_0} [\mathbf{K}(k) - \mathbf{E}(k)]
$$

It's in our best interest to work backwards: rather than integrate $T(E)$ in all its elliptical integral hairiness, we'll just differentiate this $S(E)$ expression and see that our previously-found $T(E)$ expression falls out. 

By the product rule:

$$
\frac{d}{dE} \sqrt{E+ V_0} = \frac{1}{2 \sqrt{E+V_0}}
$$

and then:

$$
\begin{aligned}
\frac{dS(E)}{dE} &= \frac{8 \sqrt{2m}}{\alpha} [(\mathbf{K}(k) - \mathbf{E}(k)) \frac{1}{2 \sqrt{E+V_0}} \\
&+ \sqrt{E+ V_0} \frac{d}{dE} (\mathbf{K}(k) - \mathbf{E}(k))]
\end{aligned}
$$

Let's get all these derivatives with respect to $k$:

$$
\frac{dk}{dE} = \frac{V_0}{k (E+ V_0)^2}, \quad \frac{d}{dE} = \frac{dk}{dE}\frac{d}{dk}
$$

It will then be useful to get the derivative of the elliptical integral difference in terms of $k$. 

$$
\begin{aligned}
\frac{d}{dE} [\mathbf{K}(k) - \mathbf{E}(k)] &= \frac{k \mathbf{E}(k)}{1-k^2} \frac{dk}{dE} \\
&= \frac{V_0\mathbf{E}(k)}{(1-k^2)(E + V_0^2)} 
\end{aligned}
$$

We can then use $1 -k^2 = \frac{2V_0}{V_0+E}$ to recude this derivative like so:

$$
\frac{d}{dE} [\mathbf{K}(k) - \mathbf{E}(k)] = \frac{\mathbf{E}(k)}{2 (E+ V_0)} 
$$




$$
\begin{aligned}
\frac{dS(E)}{dE} = \frac{8 \sqrt{2m}}{\alpha} \left[(\mathbf{K}(k) - \mathbf{E}(k)) \frac{1}{2 \sqrt{E+V_0}} + \sqrt{E+ V_0}\frac{\mathbf{E}(k)}{2 (E+ V_0)} \right]
\end{aligned}
$$

We can factor out a $\frac{1}{2 \sqrt{E+V_0}}$:

$$
\begin{aligned}
\frac{dS(E)}{dE} = \frac{8 \sqrt{2m}}{\alpha}\frac{1}{2 \sqrt{E+V_0}} \left[\mathbf{K}(k) - \mathbf{E}(k)  + {\mathbf{E}(k)} \right]
\end{aligned}
$$

From which it is apparent that the $\mathbf{E}(k)$'s cancel:

$$
\begin{aligned}
\frac{dS(E)}{dE} &= \frac{8 \sqrt{2m}}{\alpha}\frac{1}{2 \sqrt{E+V_0}} \left[\mathbf{K}(k) \right] \\
&= \frac{4 \sqrt{2m}}{\alpha}\frac{1}{ \sqrt{E+V_0}} \left[\mathbf{K}(k) \right] \\
&= \frac{4 \sqrt{2m}}{\alpha}\frac{1}{ \sqrt{E+V_0}} \left[\mathbf{K}\left(\sqrt\frac{E-V_0}{E+V_0}\right) \right] \\
&\boxed{= \frac{4 \sqrt{2m}}{\alpha  \sqrt{E+V_0}} \left[\mathbf{K}\left(\sqrt\frac{E-V_0}{E+V_0}\right) \right] = T(E)}\\
\end{aligned}
$$


which is exactly what we hoped to show.
