## 1:

A formal way of solving

$$
\hbar A^{\prime \prime} + 2iA^\prime B^\prime + iA B^{\prime \prime} = 0.
$$

is to expand $A(x)$ into a power series in $\hbar$ accoridng to 

$$
A(x) = \sum_{\nu = 0}^{\infty}A_\nu(x)\hbar^\nu
$$

Using this expansion and combining like powers of $\hbar$, show that

$$
\begin{aligned}
&2 A_0^\prime B^\prime + A_0 B^{\prime \prime} =0, \\
&A_{\nu -1}^{\prime \prime} + 2i A_\nu^\prime B^\prime + i A_\nu B^{\prime \prime} = 0, \quad \nu = 1, 2, 3...
\end{aligned}
$$

Let's start by using the power series expansion of $A(x)$ and constructing its first and second derivative:

$$
A^\prime(x) = \sum_{\nu=0}^{\infty} A^\prime_\nu(x)\hbar^\nu \\
A^{\prime\prime}(x) = \sum_{\nu=0}^{\infty} A^{\prime\prime}_\nu(x)\hbar^\nu 
$$

Then
$$
\hbar A^{\prime \prime} + 2iA^\prime B^\prime + iA B^{\prime \prime} = 0.
$$

becomes:

$$
\begin{aligned}
\hbar\sum_{\nu=0}^{\infty} A^{\prime\prime}_\nu(x)\hbar^\nu  + 2i\sum_{\nu=0}^{\infty} A^\prime_\nu(x)\hbar^\nu  B^\prime + i\sum_{\nu = 0}^{\infty}A_\nu(x)\hbar^\nu B^{\prime \prime} = 0. \\
\sum_{\nu=0}^{\infty} \left(A^{\prime\prime}_\nu(x)\hbar^{\nu+1}  + 2i A^\prime_\nu(x)\hbar^\nu  B^\prime + iA_\nu(x)\hbar^\nu B^{\prime \prime}\right) = 0.
\end{aligned}
$$

Since this equation must hold for all powers of $\hbar$, we can equate the coefficients of each power of $\hbar$ to zero separately. Now consider what happens when $\nu=0$:

$$
A_0 ^{\prime \prime} \hbar + 2i A_0^\prime B^\prime + i A_0 B^{\prime \prime} = 0
$$

Since $\hbar \ll 1$, the $\hbar$ terms die and to leading order we have

$$
\boxed{2A^\prime_0B^\prime + A_0B^{\prime \prime} = 0}
$$

where I have also canceled out the $i$'s. This is the first thing we needed to prove. 

Now let's examine some other terms in our zero-sum power series. Without loss of generality let's assume $n \geq 1$. Then the term in our zero sum becomes:

$$
\hbar^{n+1} A_n^{\prime \prime}+ 2i A_n^\prime B^\prime \hbar^n + i A_n B^{\prime \prime}\hbar^n = 0 
$$

If we divide by $\hbar^n$, this is equivalent to 

$$
\hbar A_n^{\prime \prime}+ 2i A_n^\prime B^\prime  + i A_n B^{\prime \prime} = 0 
$$

We are grouping this by terms of order $\hbar$ and higher, so in order to group all the sums in the $\nu \geq 1$ case we have:

$$
\hbar\sum_{\nu=-1}^{\infty} A^{\prime\prime}_\nu(x)\hbar^\nu  + 2i\sum_{\nu=0}^{\infty} A^\prime_\nu(x)\hbar^\nu  B^\prime + i\sum_{\nu = 0}^{\infty}A_\nu(x)\hbar^\nu B^{\prime \prime} = 0.
$$

This is informal, what we really should do is shift the index on the first term and adjust it by $\hbar$:

$$
\sum_{\nu=0}^{\infty} A^{\prime\prime}_{\nu-1}(x)\hbar^\nu  + 2i\sum_{\nu=0}^{\infty} A^\prime_\nu(x)\hbar^\nu  B^\prime + i\sum_{\nu = 0}^{\infty}A_\nu(x)\hbar^\nu B^{\prime \prime} = 0.
$$


This gives us the following recurrence relation:

$$
A^{\prime\prime}_{\nu-1}(x)\hbar^\nu  + 2i A^\prime_\nu(x)\hbar^\nu  B^\prime + iA_\nu(x)\hbar^\nu B^{\prime \prime} = 0.
$$

Dividing out by $\hbar^\nu$:
$$
\boxed{A^{\prime\prime}_{\nu-1}(x)  + 2i A^\prime_\nu(x)  B^\prime + iA_\nu(x) B^{\prime \prime} = 0.}
$$

Excellent! This is the second thing we hoped to prove. 

## 2:

A quantum particle of mass $m$ and energy $E < V_0$ is trapped inside the potential well

$$
\begin{aligned}
V(x) = \begin{cases} \infty, \quad &\text{for } x \leq 0\\
0 \quad &\text{for } 0 < x < a\\
V_0 \quad &\text{for } x \geq a\\
\end{cases}
\end{aligned}
$$

where $a$ and $V_0$ are positive constants. 

(a) Compute the bound spectrum $E_n^\text{BS}$ of the particle using Bohr-Sommerfeld quantization. Notice that $E_n^\text{BS}$ is independent of $V_0$. 

First, that condition is 

$$
S(E_n) = \oint p(x)dx = 2 \pi n \hbar
$$

by eq 5.15 for $n = 1, 2...$. 

So as usual, we can write the classical momentum as 

$$
p = \sqrt{2m (E - V(x))}
$$

But inside the well, $V(x) = 0$, so

$$
p(x) = \sqrt{2mE}
$$

The turning points of this potential are $0$ and $a$, and so we can compute the action via the closed integral in phase space:

$$
S(E) = 2 \int_0^a p(x)dx = 2 \int_0^a \sqrt{2mE}dx = 2a \sqrt{2mE}.
$$

Great. Now let's set that equal to $2\pi n \hbar$ in accordance with our BS-quantization condition:

$$
S(E_n) = 2 \pi n \hbar = 2a \sqrt{2mE}
$$
$$
 \boxed{\frac{\pi^2 n^2 \hbar^2}{2ma^2} = E_n^\text{BS}}
$$

Lo and behold, $E_n^\text{BS}$ is independent of $V_0$. 

Since $V_0$ is finite, the particle has a nonzero probability to be found in the classically forbidden regime $x >a$ (tunneling). State the quantization condition satisfied by the exact energy levels $E_n^\text{exact}$. Then show that 

1. $E_n^\text{exact}$ does not depend on $V_0$, 
2. $E_n^\text{BS} \neq E_n^\text{exact}$ for any finite $V_0$, and
3. $E_n^\text{BS} \to E_n^\text{exact}$ for $V_0 \to \infty$

Let's start by writing the stationary schrodinger equation:

$$
\frac{-\hbar^2}{2m} \frac{\partial^2 \psi}{\partial x^2} = E \psi
$$

Which we know has the general solution

$$
\psi(x) = A \sin(kx) + B\cos(kx)
$$

for $k = \sqrt{\frac{2mE}{\hbar^2}}$.

Since the function must vanish at $x = 0$, we know $B= 0$and the general solution becomes

$$
\psi(x) = A \sin(kx).
$$

But for $x >a$:

$$
-\frac{\hbar^2}{2m} \frac{\partial^2 \phi}{\partial x^2} + V_0 \psi = E \psi
$$

$$
\frac{\partial^2 \psi}{\partial x^2} = \frac{(V_0 - E)2m}{\hbar^2} \psi
$$

So let's define $k^\prime = \sqrt{\frac{2m(V_0 - E)}{\hbar^2}}$. 

That way we can write the general solution as

$$
\psi(x) = De^{k^\prime x} + Fe^{- k^\prime x}
$$

for constants $D$ and $F$. 

However, the particle is in a classically forbidden region, so the wavefunction must decay:

$$
\psi(x) = Fe^{-k^\prime x}
$$

Now to match with boundary conditions, let's shift our reference point to $x = a$ and define $C = e^{-k^\prime a}$. \
Then:

$$
\psi(x) = Ce^{-k(x-a)}.
$$

Now we can take advantage of boundary conditions. At $x = a$:

$$
A \sin(ka) = C \quad \text{and} \quad A k \cos(ka) = -k^\prime C
$$

from continuity of $\psi(x)$ and its derivative at the boundary. 

We can plug $C$ into the second equation in the statement above:

$$
Ak \cos(ka) = -k^\prime A \sin (ka)
$$

Which reduces to 

$$
-\frac{k}{k^\prime } = \tan (ka).
$$

Kinda gross. But we can re-introduce our $k$ and $k^\prime$:

$$
-\frac{\sqrt{\frac{2mE}{\hbar^2}}}{\sqrt{\frac{2m(V_0 - E)}{\hbar^2}}} = \tan \left(\sqrt{\frac{2mE}{\hbar^2}}a\right)
$$

Now even though we haven't yet solved for $E_\text{exact}$, we know it _does_ depend on $V_0$. So that's bullet point one taken care of. 

Now let's solve for $E_\text{exact}$ and show that $E_\text{exact} \neq E_\text{BS}$

$$
\frac{{\frac{2mE}{\hbar^2}}}{{\frac{2m(V_0 - E)}{\hbar^2}}} = \tan^2 \left(\sqrt{\frac{2mE}{\hbar^2}}a\right)
$$

$$
\frac{2mE}{2m(V_0 - E)} = \tan^2 \left(\sqrt{\frac{2mE}{\hbar^2}}a\right)
$$

$$
\frac{E}{(V_0 - E)} = \tan^2 \left(\sqrt{\frac{2mE}{\hbar^2}}a\right)
$$

$$
{E} = V_0\tan^2 \left(\sqrt{\frac{2mE}{\hbar^2}}a\right) - E \tan^2 \left(\sqrt{\frac{2mE}{\hbar^2}}a\right).
$$

$$
E + E \tan^2 \left(\sqrt{\frac{2mE}{\hbar^2}}a\right) = V_0\tan^2 \left(\sqrt{\frac{2mE}{\hbar^2}}a\right) 
$$

$$
E\left(1 +  \tan^2 \left(\sqrt{\frac{2mE}{\hbar^2}}a\right)\right) = V_0\tan^2 \left(\sqrt{\frac{2mE}{\hbar^2}}a\right) 
$$

Now use $\sec^2 = 1+ \tan^2$:

$$
E \sec^2 \left(\sqrt{\frac{2mE}{\hbar^2}}a\right) = V_0\tan^2 \left(\sqrt{\frac{2mE}{\hbar^2}}a\right) 
$$

$$
E  = \frac{V_0\tan^2 \left(\sqrt{\frac{2mE}{\hbar^2}}a\right)}{\sec^2 \left(\sqrt{\frac{2mE}{\hbar^2}}a\right)} 
$$

Then finally, 
$$
E_\text{exact}  = V_0\sin^2 \left(\sqrt{\frac{2mE}{\hbar^2}}a\right),
$$

which is not equal to $E_n^\text{BS} = \frac{\pi^2 n^2 \hbar^2}{2ma}$. So that takes care of bullet point 2. 

Now let's take $V_0 \to \infty$:

$$
\frac{E_\text{exact}}{V_0}  = \sin^2 \left(\sqrt{\frac{2mE}{\hbar^2}}a\right). 
$$

Taking $V_0 \to 0$ means:

$$
\sin^2 \left(\sqrt{\frac{2mE}{\hbar^2}}a\right) = 0
$$

and thus, valid solutions are of the form:

$$
\sqrt{\frac{2mE}{\hbar^2}}a = n \pi \quad \text{for } n = 1, 2, 3...
$$

And thus we can solve for $E_\text{exact}$ in this limit:

$$
E_\text{exact} = \frac{\hbar^2}{2m}\left(\frac{n \pi }{a}\right)^2 = E_\text{BS}
$$

And that accomplishes bullet point 3. Great!

## 3:

Show that WKB quantization according to 5.49 yields the exact result for energy levels of the harmonic oscillator (eq 4.55)

First, let's write the form of that quantization:

$$
S(E_n) = \oint p(x)dx= 2\pi \hbar \left(n - \frac{1}{2}\right)
$$

First we need to calculate $S(E_n)$ for a harmonic oscillator. We know that $T(E) = \frac{2\pi}{\omega}$ in this case, so we can simply integrate to find $S(E_n)$:

$$
S(E) = \int_0^S dS(E) = \int_0^ET(E)dE = \frac{2\pi}{\omega}E.
$$

Then in accordance with the WKB quantization condition:

$$
\frac{2\pi}{\omega}E  = 2\pi \hbar \left(n - \frac{1}{2}\right)
$$

for $n = 1, 2, 3...$.

Then we simply rearrange:

$$
\boxed{{E_n}  = \hbar\omega \left(n - \frac{1}{2}\right)}
$$

This is indeed the energy levels of the harmonic oscillator. Only catch catch is that Eq. 4.55 uses zero-indexing, and minus in the parenthetical term would go to a plus as a result of the shift.

Again, let's start by writing the WKB quantization condition:

$$
S(E_n) = 2\pi \hbar\left(n - \frac{1}{2}\right)
$$

For the hypercosine potential, we proved in HW 2 that the accumulated action is equal to 

$$
S(E) = \frac{8 \sqrt{2m}}{\alpha} \sqrt{E + V_0} \left[\mathbf{K}\left(\sqrt\frac{E-V_0}{E+V_0}\right) - \mathbf{E} \left(\sqrt{\frac{E-V_0}{E+V_0}}\right) \right]
$$

So first, expand in terms of $k$, and kill all terms of higher order than quadratic in $k$. 

From Eq 2.40 and 2.41, the elliptical integrals are approximated thusly:

$$
\mathbf{K}(k) \approx \frac{\pi}{2} \left(1 + \frac{1}{4}k^2\right)
$$

$$
\mathbf{E}(k) \approx \frac{\pi}{2} \left(1 - \frac{1}{4}k^2\right)

The difference is then:

$$
\mathbf{K}(k) - \mathbf{E}(k) \approx \frac{\pi}{4}k^2
$$

Which means our accumulated action becomes:

$$
S(E) \approx \frac{8 \sqrt{2m}}{\alpha} \sqrt{E + V_0} \left(\frac{\pi}{4}\right)\left[\frac{E-V_0}{E+ V_0}\right].
$$

This reduces to:

$$
S(E) \approx \frac{2 \pi \sqrt{2m}}{\alpha}  \left[\frac{E-V_0}{\sqrt{E+ V_0}}\right].
$$

But this is still not enough. Fortunately we care only about low-lying energy states, and so we can approximate

$$
E_n  = V_0+ \epsilon_n, \quad \text{for } \epsilon_n \ll V_0.
$$

Then under this limit:

$$
\sqrt{E+ V_0 } \approx \sqrt{2V_0 + \epsilon_n} \approx \sqrt{2V_0}
$$

And our action is then

$$
\begin{aligned}
S(E) &\approx \frac{2 \pi \sqrt{2m}}{\alpha}  \left[\frac{E-V_0}{\sqrt{2V_0}}\right] \\
&= \frac{2 \pi \sqrt{m}}{\alpha}  \left[\frac{\epsilon_n}{\sqrt{V_0}}\right] \\
\end{aligned}
$$

Then we use our WKB quantization condition...
$$
\frac{2\pi \epsilon_n}{\alpha} \sqrt \frac{m}{V_0}  = 2 \pi \hbar \left(n - \frac{1}{2}\right)
$$

where I have simplified $S$ a little bit on the LHS. We can then solve for  $\epsilon_n$:

$$
{ \epsilon_n}   =  \alpha\hbar \left(n - \frac{1}{2}\right)\sqrt{\frac{V_0}{m}}
$$

And from there simply add back in $V_0$ to get the low-lying energy levels:

$$
\boxed{E^\text{WKB}_n \approx V_0 + \hbar \alpha \sqrt{\frac{V_0}{m}} \left(n - \frac{1}{2}\right)}
$$

Great! Now let's see if it passes the H.O. test. To begin, shift it so that the energy-levels are zero-indexed for comparison:

$$
E^\text{WKB}_n \approx V_0 + \hbar \alpha \sqrt{\frac{V_0}{m}} \left(n + \frac{1}{2}\right) \quad \text{for } n = 0, 1, 2...
$$

For a potential with a quadratic minimum at $x_0$, the H.O. test is passed when the following condition is met:

$$
E^\text{WKB}_n \approx E^\text{osc}_n  \quad \text{for } n = 0, 1, 2...
$$

where 

$$
E^\text{osc}_n = V(x_0 ) + \hbar \omega \left(n + \frac{1}{2}\right),
$$

and 

$$
\omega = \sqrt{\frac{V^{\prime \prime}(x_0)}{m}}.
$$

So let's find the point $x_0$ where our potential has that quadratic minimum. This is true when $V^{\prime}(x_0) =0$ and $V^{\prime\prime}(x_0) > 0$. 

The first derivative of the hypercosine potential is then:

$$
\frac{d}{dx}\left[V_0 \cosh (\alpha x)\right] = \alpha V_0 \sinh (\alpha x)
$$

Setting this to zero reveals that the only point this first derivative could be zero is at $\alpha = 0$. Thankfully it has positive curvature, which we can easily check:

$$
\frac{d^2}{dx^2}\left[V_0 \cosh (\alpha x)\right] = \alpha^2 V_0 \sinh (\alpha x)
$$

Which is clearly positive at $x =0$. 

So let's now evaluate $\omega$:

$$
\omega = \sqrt{\frac{\alpha^2 V_0}{m}}.
$$

And then we plug that back into 

$$
\begin{aligned}
E^\text{osc}_n &= V(x_0 ) + \hbar \omega \left(n + \frac{1}{2}\right) \\
&= \boxed{V(x_0 ) + \hbar \alpha \omega\sqrt{\frac{ V_0}{m}} \left(n + \frac{1}{2}\right)}
\end{aligned}
$$

Aha! This is the same expression we found for $E_n$'s WKB approximation, and hence, our WKB approximation passes the H.O. test.

```python
import numpy as np
from scipy.special import ellipk
import matplotlib.pyplot as plt
levels = np.arange(0.0, 6.0, 1.)
def e(n):
    return (1/ellipk(-1) * 3/(4 * np.sqrt(2)) * np.pi * (n + 1/2))**(4/3)
```

```python
wkb_approx = e(levels)
wkb_exact = np.array([0.66798, 2.39364, 4.69679, 7.33573, 10.24430, 13.37933])
```

We define the classical period of motion as:

$$
T(E) = \oint \frac{dx}{\sqrt{\frac{2}{m}(E - V(x))}}
$$

where the integral is taken over one full oscillation from $x_{\text{turn}}^-$ to $x_{\text{turn}}^+$ and back.

We have the energy equation:

$$
E = V_0 \left(\frac{1}{(x \alpha)^2} + (x \alpha)^2\right),
$$

which rearranges to

$$
\begin{aligned}
E - V(x) &= V_0 \left(\frac{1}{(x\alpha)^2} + (x\alpha)^2 \right) - V_0 \left(\frac{1}{(x\alpha)^2} + (x\alpha)^2 \right) \\
&= V_0 \left(\frac{E}{V_0} - \frac{1}{(x\alpha)^2} - (x\alpha)^2 \right)
\end{aligned}
$$

Introducing $u = x \alpha$, the integral transforms into:

$$
T(E) = \sqrt{2m} \int \frac{dx}{\sqrt{E - V(x)}}
$$

$$
= \sqrt{2m} \int \frac{du}{\alpha \sqrt{V_0 \left(\frac{E}{V_0} - \frac{1}{u^2} - u^2 \right)}}.
$$

Factoring out $V_0$, we get:

$$
T(E) = \sqrt{\frac{m}{2V_0}} \int \frac{du}{\alpha \sqrt{\frac{E}{V_0} - \frac{1}{u^2} - u^2}}.
$$

We recognize the integral as a standard form:

$$
\int \frac{dx}{\sqrt{a + bx + cx^2}} = - \frac{1}{\sqrt{|c|}} \arcsin\left(\frac{2cx + b}{\sqrt{|\Delta|}}\right),
$$

valid for $\Delta = 4ac - b^2 < 0$ and $c < 0$.

Here, we match terms:

$$
a = \frac{E}{V_0}, \quad b = 0, \quad c = -1.
$$

The discriminant condition:

$$
\Delta = 4ac - b^2 = 4 \left(\frac{E}{V_0} \right)(-1) - 0 = -4 \frac{E}{V_0} < 0,
$$

is satisfied for $E > 0$.

Using the known result:

$$
T(E) = \sqrt{\frac{m}{2V_0}} \frac{1}{\alpha \sqrt{1}} \left( \pi \right),
$$

we simplify to:

$$
T(E) = \frac{\pi}{\alpha} \sqrt{\frac{m}{2V_0}}.
$$

Thus, the final result for the period of motion is:

$$
T(E) = \frac{\pi}{\alpha} \sqrt{\frac{m}{2V_0}}.
$$

All we need to do is integrate with respect to $E$:

$$
S(E) = \int_{E_\text{min}}^{E_\text{max}}T(E)dE
$$

We found $E_\text{min} = 2V_0$ in a previous part, and $E_\text{max}$ must be the Hamiltonian by conservation of energy. 

$$
\begin{aligned}
S(E) &= \int_{2V_0}^{E}T(E)dE \\
&= \int_{2V_0}^{E}\frac{\pi}{\alpha} \sqrt{\frac{m}{2V_0}}dE \\
&= \boxed{\frac{\pi}{\alpha} \sqrt{\frac{m}{2V_0}}(E - 2V_0)}
\end{aligned}
$$

(e) From the actiona integral we can use the WKB formula to figure out the energy levels explicitly:

$$
\begin{aligned}
S(E) &= 2 \pi\hbar \left(n - \frac{1}{2}\right) \\
\frac{\pi}{\alpha} \sqrt{\frac{m}{2V_0}}(E - 2V_0) &= 2 \pi\hbar \left(n - \frac{1}{2}\right)
\end{aligned}
$$

And then we can solve for $E$:

$$
\sqrt{\frac{m}{2V_0}}(E - 2V_0) = 2 \alpha \hbar \left(n - \frac{1}{2}\right)
$$

$$
\sqrt{\frac{m}{2V_0}}(E)  = 2 \alpha \hbar \left(n - \frac{1}{2}\right) + \sqrt{\frac{m}{2V_0}}( 2V_0)
$$
