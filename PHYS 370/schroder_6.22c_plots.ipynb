```python
import numpy as np
import matplotlib.pyplot as plt
M= []
b = np.arange(0.01, 5, 0.01)
two_j_max = 10
for twoj in range(1, two_j_max, 2):
    plt.plot(
       b, ((twoj/2 + 1/2))/(np.tanh(b*(twoj/2+1/2))) - 1/2 * 1/(np.tanh(b/2)), 
        label = 'j = ' + str(twoj/2)
    )
plt.xlabel('b', size = 16)
plt.ylabel(r'$\frac{MN} {\delta_u}$', size = 20)
plt.legend()
plt.savefig('6.22_c_plots.jpg')
```
